### **Paso 2.2 - Ingesta del archivo "races.csv"**

#### Paso 1 - Leer el archivo CSV

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType
```

```python
races_schema = StructType(fields=[StructField("raceId", IntegerType(), False),
                                  StructField("year", IntegerType(), True),
                                  StructField("round", IntegerType(), True),
                                  StructField("circuitId", IntegerType(), True),
                                  StructField("name", StringType(), True),
                                  StructField("date", DateType(), True),
                                  StructField("time", StringType(), True),
                                  StructField("url", StringType(), True) 
])
```

```python
races_df = spark.read \
.option("header", True) \
.schema(races_schema) \
.csv("dbfs:/mnt/formula1dl/raw/races.csv")
```

```python
races_df.show(truncate=False)
```

#### Paso 2 - Añadir las columnas "ingestion_date" y "race_timestamp"

```python
from pyspark.sql.functions import to_timestamp, concat, col, lit, current_timestamp
```

```python
races_with_timestamp_df = races_df.withColumn("ingestion_date", current_timestamp()) \
                                  .withColumn("race_timestamp", to_timestamp(concat(col('date'), lit(' '), col('time')), 'yyyy-MM-dd HH:mm:ss'))
```

```python
races_with_timestamp_df.show(truncate=False)
```

#### Paso 3 - Seleccionar sólo las columnas necesarias y renombrarlas como corresponda

```python
races_selected_df = races_with_timestamp_df.select(col('raceId').alias('race_id'), col('year').alias('race_year'), 
                                                   col('round'), col('circuitId').alias('circuit_id'),col('name'), col('ingestion_date'), col('race_timestamp'))
```

```python
races_selected_df.show(truncate=False)
```

#### Paso 4 - Escribir datos en el contenedor **processed** del ADLS como **parquet**

```python
# Escribimos el archivo con formato PARQUET en el contenedor "processed" y directorio "races"
races_selected_df.write.mode('overwrite').parquet("/mnt/formula1dl/processed/races")
```

```python
# Visualizamos los objetos que se encuentran en el directorio "races"
%fs
ls /mnt/formula1dl/processed/races
```

```python
# Podemos particionar el archivo al momento de escribirlo de acuerdo a los distintos años en la columna "race_year"
races_selected_df.write.mode('overwrite').partitionBy('race_year').parquet("/mnt/formula1dl/processed/races")
```
