### **Agregar variables de entorno**

Podemos añadir variables de entorno de dos formas:

- Utilizando **ENV** en el **Dockerfile**
- Modificando el archivo **/etc/environment**

Ambos métodos, utilizando `ENV` en el Dockerfile y modificando el archivo `/etc/environment`, son válidos para establecer variables de entorno en un contenedor Docker basado en Ubuntu. La elección entre ellos dependerá de tus preferencias y requisitos específicos.

Por ejemplo, pensemos que necesito crear la variable de entorno PATH,Aaquí hay algunas consideraciones para cada enfoque:

1. **Utilizando `ENV` en el Dockerfile**:
   - **Ventajas**:
     - Es específico para el contenedor Docker y se aplica cuando se crea y se ejecuta el contenedor.
     - Es fácil de configurar y mantener dentro del Dockerfile.
   - **Desventajas**:
     - Los cambios en la variable de entorno requieren la reconstrucción de la imagen del contenedor si deseas aplicarlos.
     - Puede ser menos flexible si necesitas modificar la variable de entorno después de que el contenedor esté en ejecución.

2. **Modificando el archivo `/etc/environment`**:
   - **Ventajas**:
     - Los cambios en el archivo `/etc/environment` se aplican a nivel de sistema y se reflejan en todos los procesos del contenedor, no solo en los procesos iniciados por el entorno Docker.
     - Los cambios son persistentes y se mantienen incluso después de que el contenedor se reinicie.
   - **Desventajas**:
     - Puedes necesitar acceso root dentro del contenedor para modificar el archivo `/etc/environment`, lo que podría complicar la configuración en entornos automatizados.
     - Pueden surgir conflictos si otros procesos o aplicaciones dentro del contenedor dependen de una configuración diferente de `PATH`.

En general, si tus necesidades son simples y no esperas cambios frecuentes en la variable de entorno `PATH`, es conveniente utilizar `ENV` en el Dockerfile para mantener la configuración de manera clara y controlada. Sin embargo, si necesitas flexibilidad y persistencia en la configuración de la variable de entorno `PATH`, modificar el archivo `/etc/environment` puede ser una mejor opción.

### **Comandos importantes**

#### **`SSH`**

```python
# Verificar que SSH este ejecutandose en el contenedor
service ssh status
```

```python
# Para entrar a un nodo x desde otro nodo
ssh <host_name o id del nodo>
```

```python
# Copiar archivos entre nodos utilizando SSH (Ejemplo: copia de un archivo desde el namenode a un nodo)
scp core-site.xml dnnm1:/opt/bd/hadoop/etc/hadoop
```

#### **`HDFS`**

```python
# Nos va a listar todos los procesos de JAVA que esten en ejecución, cuando lanzemos un "namenode" por ejemplo y nos debe devolver 
# la palabra "Namenode" y "Jps"
jps
```

```python
# Nos sirve para visualizar las estadisticas y el estado del sistema de archivos distribuido (HDFS)
# Nos saldrá todo en 0 si es que no hemos arracando ningun datanode
# Este comando podemos ejecutarlo tanto desde el namenode como de un datanode
hdfs dfsadmin -report
```

```python
# Visualizar la memoria en disco que tiene nuestro nodo
hdfs dfs -df -h
```

```python
# Si modificamos el archivo de configuracion (hdfs-site.xml) y queremos ejecutar los cambios inmediatamente sin detener y 
# volver a ejecutar el nodo
hdfs dfsadmin -reconfig namenode namenode:9000 start
```

```python
# Para comprobar una propiedad del archivo del archivo "hdfs-site.xml", por ejemplo, la propiedad "dfs.heartbeat.interval", 
# lo hacemos de la siguiente manera:
hdfs getconf -confKey dfs.heartbeat.interval
```

```python
# Si modificamos un archivo dentro del NameNode, como por ejemplo, el archivo "workers" podemos apagar y volver a encender 
# el nodo, o podemos ejecutar el siguiente comando:
hdfs dfsadmin -refreshNodes
```

```python
# Para comprobar que todos los archivos de configuracion de la ruta ${HADOOP_HOME}/etc/hadoop esten correctos (no es confiable al 100%)
hadoop conftest
```

#### **`YARN`**

```python
# Visualizar el estado del ResourceManager
yarn rmadmin -getServiceState rm
```

```python
# Verificar el estado de los nodos del clúster
yarn node -list
```

```python
# Verificar el estado de las aplicaciones en ejecución
yarn application -list
```

```python
# Verificar el estado de los trabajos en ejecución
yarn application -appStates RUNNING -list
```
