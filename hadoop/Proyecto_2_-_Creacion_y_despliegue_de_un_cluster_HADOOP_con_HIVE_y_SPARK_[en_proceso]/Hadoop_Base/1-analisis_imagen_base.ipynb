### **`Dockerfile`**

**`ENV LOG_TAG "[BASE Hadoop_${HADOOP_VERSION}]:"`**: Aquí se define la variable **LOG_TAG** con un valor que parece ser un tag para los logs. Utiliza la variable **HADOOP_VERSION** definida anteriormente para incluir la versión de Hadoop en el tag.

En este caso, corresponde a la URL de un repositorio de distribución de Hadoop.

````python
# Establecemos las variables
ENV HADOOP_VERSION 3.3.6
ENV BASE_DIR /root
ENV LOG_TAG "[BASE Hadoop_${HADOOP_VERSION}]:"
ENV REPOSITORY https://dlcdn.apache.org/hadoop/common
````

**`apt-get update && \`**: Actualiza la lista de paquetes disponibles para su descarga.

**`apt-get install -y --no-install-recommends openssh-server openjdk-8-jdk wget python3 curl locales && \`**: Esta línea de comando se utiliza en sistemas Linux basados en Debian (como Ubuntu) para instalar una lista de paquetes utilizando el administrador de paquetes apt-get. Aquí está el desglose:

- `apt-get install`: Este comando indica al administrador de paquetes apt-get que instale los paquetes especificados a continuación.

- `-y`: Esta opción indica a apt-get que responda "sí" a todas las preguntas de confirmación automáticamente. Es útil para automatizar la instalación de paquetes en scripts sin necesidad de intervención manual.

- `--no-install-recommends`: Esta opción le dice a apt-get que no instale automáticamente los paquetes recomendados como dependencias. Esto significa que solo se instalarán los paquetes que se han especificado explícitamente, y no se instalarán otros paquetes recomendados que normalmente se instalarían junto con los paquetes especificados.

- `openssh-server`: Este es el paquete que instala el servidor SSH en el sistema. El servidor SSH permite conexiones remotas seguras al sistema a través del protocolo SSH. El paquete de software "**openssh-server**" proporciona un servidor SSH en sistemas basados en Unix/Linux, como Ubuntu. `OpenSSH` es una implementación libre y de código abierto del protocolo SSH (Secure Shell), que proporciona cifrado seguro de datos de comunicación, autenticación de claves y otras características relacionadas con la seguridad. El paquete "**openssh-server**" específicamente instala el servidor SSH en un sistema Ubuntu, permitiendo que el sistema actúe como un servidor SSH al que otros usuarios o sistemas pueden conectarse de forma segura.Cuando instalas el paquete "**openssh-server**" en un sistema Ubuntu, esencialmente estás instalando el software necesario para habilitar la funcionalidad del servidor SSH, lo que te permite iniciar y administrar conexiones SSH entrantes hacia ese sistema. Este paquete es muy común y esencial en muchos entornos de servidor y desarrollo, ya que SSH es una herramienta fundamental para administrar sistemas de forma remota de manera segura.

- `openjdk-8-jdk`: Este es el paquete que instala el kit de desarrollo de Java (JDK) versión 8 en el sistema. Es necesario para desarrollar y ejecutar aplicaciones Java en el sistema.

- `python3`: Este es el paquete que instala la versión 3 del intérprete de Python en el sistema. Python es un lenguaje de programación popular utilizado para una variedad de propósitos, incluyendo scripting, desarrollo web, análisis de datos y más.

- `curl`: Este es el paquete que instala la utilidad curl, que se utiliza para transferir datos desde o hacia un servidor utilizando uno de los protocolos compatibles, como HTTP, HTTPS, FTP, FTPS, SCP, SFTP, TFTP, LDAP, LDAPS, DICT, FILE, IMAP, SMTP, POP3, RTSP, RTMP y más.

- `locales`: Este es el paquete que instala archivos de configuración para el soporte de locales en el sistema. Los locales determinan las preferencias de idioma, formato de fecha y hora, formato de moneda y otras configuraciones regionales en el sistema.

- En resumen, esta línea de comando instala una lista de paquetes esenciales en el sistema, incluyendo el servidor SSH, el JDK de Java 8, utilidades de descarga de archivos, el intérprete de Python 3, la utilidad curl y archivos de configuración para soporte de locales.

**`apt-get clean && \`**: Limpia la caché de **apt-get** para reducir el tamaño de la imagen final.

**`rm -rf /var/lib/apt/lists`**: Elimina los archivos de lista de paquetes descargados, lo que también ayuda a reducir el tamaño de la imagen final y elimina información no necesaria.

````python
#Instalamos el software necesario para Hadoop
#Instalamos un servidor SSH, JDK8, wget y Python
RUN echo "$LOG_TAG actualizando e instalando paquetes básicos" && \
    apt-get update && \
    apt-get install -y --no-install-recommends openssh-server openjdk-8-jdk wget python3 curl locales && \ 
    apt-get clean && \
    rm -rf /var/lib/apt/lists    
````

El comando **`RUN`** ejecuta una serie de comandos en el contenedor durante la construcción de la imagen:

**`echo "$LOG_TAG Descargando Hadoop"`**: Imprime un mensaje log que utiliza el valor de la variable de entorno LOG_TAG para indicar que se está descargando Hadoop.

**`wget "${REPOSITORY}/hadoop-${HADOOP_VERSION}/hadoop-${HADOOP_VERSION}.tar.gz"`**: Descarga el archivo tar.gz de la versión especificada de Hadoop desde el repositorio definido en la variable de entorno REPOSITORY.

**`ENV JAVA_HOME /usr/lib/jvm/java-8-openjdk-amd64/`**: Esta línea establece la variable de entorno **JAVA_HOME** con la ruta al directorio de instalación de Java en el contenedor.

**`ENV HADOOP_HOME=/usr/local/hadoop`**: Esta línea establece la variable de entorno **HADOOP_HOME** con el valor **/usr/local/hadoop**. **HADOOP_HOME** representa la ubicación del directorio principal de instalación de Hadoop en el contenedor Docker. 

**`ENV PATH ${PATH}:${HADOOP_HOME}/bin/:${HADOOP_HOME}/sbin/`**: Esta línea modifica la variable de entorno **PATH** para incluir los directorios donde residen los binarios de Hadoop (`${HADOOP_HOME}/bin/` y `${HADOOP_HOME}/sbin/`). Esto permite que los comandos de Hadoop estén disponibles en la ruta de búsqueda del sistema.

````python
#Establecemos las variables de entorno JAVA_HOME apuntando a la ruta del JDK, 
#HADOOP_HOME apuntando a la ruta en la que hemos copiado Hadoop e incluimos en 
#la variable PATH las rutas con los binarios y scripts para gestionar Hadoop.
RUN echo "$LOG_TAG Establecer variables de entorno"
ENV JAVA_HOME=/usr/lib/jvm/java-8-openjdk-amd64 
ENV HADOOP_HOME=/usr/local/hadoop
ENV PATH=$PATH:${HADOOP_HOME}/bin:${HADOOP_HOME}/sbin 
````

Estas líneas de comando se utilizan en un archivo Dockerfile para realizar acciones específicas dentro de un contenedor Docker durante el proceso de creación de la imagen. Aquí está el desglose:

1. **`RUN ssh-keygen -t rsa -f ~/.ssh/id_rsa -P '' && \`**

    - `RUN`: Esta instrucción indica a Docker que ejecute un comando dentro del contenedor mientras se está construyendo la imagen.
    - `ssh-keygen`: Es el comando utilizado para generar claves SSH.
    - `-t rsa`: Especifica el tipo de clave que se generará como RSA.
    - -`f ~/.ssh/id_rsa`: Especifica la ubicación y el nombre del archivo de la clave privada que se generará. En este caso, se almacenará en el directorio `~/.ssh/` (el directorio de inicio del usuario, en nuestro caso **/root**) con el nombre **id_rsa**.
    - `-P ''`: Define una frase de contraseña en blanco para evitar que se solicite una interacción durante la generación de la clave. Una "frase de contraseña" se refiere a una secuencia de caracteres que se utiliza para proteger la clave privada generada. Cuando se crea una clave SSH, se puede asociar una frase de contraseña con la clave privada para proporcionar una capa adicional de seguridad. Esta frase de contraseña actúa como una clave de acceso adicional, además de la propia clave privada, y se solicitará cada vez que se utilice la clave privada para autenticarse con un servidor SSH. En el contexto del comando **ssh-keygen**, la opción **-P** permite especificar la frase de contraseña para la clave privada. Al pasar una cadena vacía (**''**) como argumento para **-P**, se indica que no se desea utilizar una frase de contraseña. Esto significa que la clave privada generada estará sin protección por contraseña y no se solicitará una contraseña cuando se utilice para autenticarse con servidores SSH. Es importante tener en cuenta que mientras que omitir una frase de contraseña puede hacer que el proceso de autenticación sea más conveniente, también implica un riesgo de seguridad adicional, ya que cualquiera que obtenga acceso a la clave privada podrá utilizarla sin necesidad de una contraseña adicional. Por lo tanto, es una compensación entre comodidad y seguridad, y debe considerarse cuidadosamente según el contexto de uso.

    - `&& \`: Este operador se utiliza para concatenar comandos en una sola línea en un archivo Dockerfile. Indica que el siguiente comando se ejecutará si el anterior tiene éxito.

Aquí hay un desglose detallado:

    - `cat`: Es un comando en Unix/Linux que se utiliza para concatenar archivos y mostrar su contenido en la salida estándar.

    - `~/.ssh/id_rsa.pub`: Es la ubicación del archivo de clave pública. Este es el archivo que contiene la clave pública generada previamente. La convención `~` representa el directorio de inicio del usuario.

    - `>>`: Es un operador de redirección en Unix/Linux que se utiliza para agregar la salida de un comando a un archivo existente.

    - `~/.ssh/authorized_keys`: Es la ubicación del archivo **authorized_keys**. Este archivo contiene las claves públicas autorizadas para autenticación SSH en el servidor. Se encuentra típicamente en el directorio `~/.ssh/` del usuario y es utilizado por el servidor SSH para permitir el acceso a usuarios autorizados.

Esto permite que la clave pública sea utilizada para autenticar el acceso SSH al contenedor cuando se conecta al mismo. Es una práctica común para permitir el acceso SSH sin necesidad de introducir la contraseña en cada conexión.

**`¿Qué es SSH?`**

`SSH`, que significa "Secure Shell" (shell seguro), **es un protocolo de red que permite a los usuarios acceder de forma segura a un `sistema remoto` a través de una conexión cifrada. Proporciona una forma segura de iniciar sesión en un sistema remoto, transferir archivos de forma segura y ejecutar comandos de forma remota**.

Aquí hay algunas características clave de SSH:

- `Cifrado`: SSH cifra toda la comunicación entre el cliente y el servidor, lo que proporciona privacidad y seguridad para los datos transmitidos, incluidas las credenciales de inicio de sesión y cualquier otra información confidencial.

- `Autenticación`: SSH proporciona varios métodos de autenticación, incluidos la autenticación de contraseña, la autenticación basada en clave pública y la autenticación basada en certificados. Estos métodos ayudan a garantizar que solo los usuarios autorizados puedan acceder al sistema remoto.

- `Integridad de datos`: SSH utiliza técnicas para garantizar que los datos transmitidos no sean alterados durante la transferencia, lo que ayuda a proteger contra ataques de manipulación de datos.

- `Reenvío de puertos`: SSH permite el reenvío seguro de puertos, lo que facilita el acceso a servicios de red en sistemas remotos a través de una conexión segura.

- `Túneles seguros`: SSH puede crear túneles seguros a través de redes no seguras, lo que permite el acceso seguro a recursos internos de una red desde ubicaciones externas.

En resumen, SSH es una herramienta fundamental en la administración de sistemas informáticos, ya que proporciona una forma segura y cifrada de administrar sistemas remotos y transferir datos de forma segura a través de una red.

**`¿A que se refiere con sistema remoto?`**

Cuando se habla de un "**sistema remoto**", se hace referencia a un sistema informático que está ubicado en un lugar diferente al que se encuentra el usuario que desea acceder a él. Este sistema remoto puede ser cualquier dispositivo, como un servidor, una computadora personal, un enrutador, un dispositivo IoT (Internet de las cosas), entre otros, que está conectado a una red, ya sea una red local (LAN) o una red más amplia, como Internet.

La característica principal de un sistema remoto es que se puede acceder y controlar desde una ubicación distinta a la que ocupa físicamente. Esto se logra a través de redes de comunicación, como Internet, utilizando protocolos como SSH, que permiten una conexión segura y cifrada entre el usuario y el sistema remoto.

Algunos ejemplos de situaciones en las que se puede acceder a un sistema remoto incluyen:

- `Administración de servidores`: Los administradores de sistemas pueden acceder a servidores remotos para realizar tareas de administración, como la configuración, la instalación de software, el monitoreo del rendimiento y la solución de problemas.

- `Trabajo remoto`: Los empleados pueden acceder a sus computadoras de trabajo desde su casa o desde cualquier otro lugar fuera de la oficina para realizar tareas laborales, acceder a archivos y aplicaciones, y colaborar con colegas.

- `Acceso a dispositivos IoT`: Los propietarios de dispositivos IoT, como cámaras de seguridad, termostatos inteligentes y dispositivos de automatización del hogar, pueden acceder a ellos de forma remota para controlar su funcionamiento y ajustar la configuración.

En resumen, un sistema remoto es simplemente un sistema informático que se puede acceder y controlar de manera remota a través de una red de comunicación, lo que permite una amplia gama de aplicaciones y casos de uso en diversos entornos.

````python
#Creamos las claves ssh necesarias para que se comunique el nodo master con los nodos slaves. Al tener todos las 
#mismas claves e incluirlas como claves autorizadas, el acceso ssh entre los distintos nodos está garantizado.
RUN echo "$LOG_TAG Crear claves SSH" && \
    ssh-keygen -t rsa -f ~/.ssh/id_rsa -P '' && \
    cat ~/.ssh/id_rsa.pub >> ~/.ssh/authorized_keys
````

**`mkdir -p`**: Esto es parte del comando **mkdir**, que se utiliza para crear directorios en sistemas Unix. La opción **-p** indica que se deben crear todos los directorios necesarios en la ruta especificada, incluso si los directorios padres aún no existen.

**`${BASE_DIR}`**: Esto es una variable de entorno que se está utilizando para especificar la ruta del directorio que se creará. En el Dockerfile, esta variable debe fue definida anteriormente con una ruta específica, **/root**.

**`/hdfs/namenode`** y **`/hdfs/datanode`**: Estas son las rutas de directorios que se crearan.

**`mkdir ${HADOOP_HOME}/logs`**: Ejecuta el comando `mkdir` para crear un directorio llamado `logs` dentro del directorio especificado por la variable de entorno `HADOOP_HOME`.

- `/tmp/`: Es el directorio de destino dentro del sistema de archivos del contenedor donde se copiarán los archivos seleccionados. En este caso, los archivos seleccionados se copiarán al directorio /tmp/ dentro del contenedor.

En resumen, estas líneas de comando en el Dockerfile copian todos los archivos y directorios dentro del directorio Config-files del sistema de archivos del host hacia el directorio /tmp/ dentro del sistema de archivos del contenedor durante el proceso de construcción de la imagen. Estas líneas de comando en un Dockerfile mueven archivos desde una ubicación específica dentro del sistema de archivos del contenedor hacia otras ubicaciones dentro del mismo sistema de archivos. Aquí está el desglose:

- `RUN`: Es una instrucción de Dockerfile que se utiliza para ejecutar comandos durante el proceso de construcción de la imagen del contenedor.

- `mv`: Es el comando utilizado en sistemas Unix/Linux para mover archivos o directorios de una ubicación a otra.

- `/tmp/start-hadoop.sh`, `/tmp/hadoop-env.sh`, `/tmp/hdfs-site.xml`, `/tmp/core-site.xml`, `/tmp/mapred-site.xml`, `/tmp/yarn-site.xml`, `/tmp/slaves`, `/tmp/ssh_config`: Son las ubicaciones de los archivos dentro del sistema de archivos del contenedor que se están moviendo.

- `${BASE_DIR}/start-hadoop.sh`, `/usr/local/hadoop/hadoop-env.sh`, `/usr/local/hadoop/hdfs-site.xml`, `/usr/local/hadoop/core-site.xml`, `/usr/local/hadoop/mapred-site.xml`, `/usr/local/hadoop/yarn-site.xml`, `/usr/local/hadoop/slaves`, `${BASE_DIR}/.ssh/config`: Son las ubicaciones de destino dentro del sistema de archivos del contenedor hacia donde se están moviendo los archivos.

Esto es comúnmente utilizado para configurar el entorno y la aplicación dentro del contenedor.

````python
RUN echo "$LOG_TAG Mover archivos" && \
    mv /tmp/start-hadoop.sh ${BASE_DIR}/start-hadoop.sh && \
    mv /tmp/hadoop-env.sh /usr/local/hadoop/hadoop-env.sh && \
    mv /tmp/hdfs-site.xml /usr/local/hadoop/hdfs-site.xml && \ 
    mv /tmp/core-site.xml /usr/local/hadoop/core-site.xml && \
    mv /tmp/mapred-site.xml /usr/local/hadoop/mapred-site.xml && \
    mv /tmp/yarn-site.xml /usr/local/hadoop/yarn-site.xml && \
    mv /tmp/slaves /usr/local/hadoop/slaves && \
    mv /tmp/ssh_config ${BASE_DIR}/.ssh/config 
````

### **Config-files/`ssh_config`**

El archivo **`ssh_config`** es un archivo de configuración para el cliente SSH en sistemas Unix y Linux. Controla el comportamiento de las conexiones SSH salientes desde el sistema en el que se encuentra el archivo.
```
Host localhost
  StrictHostKeyChecking no
```
Esto configura una opción específica para las conexiones SSH que se hacen al **host localhost**.

- **`Host localhost`**: Esto especifica que las siguientes opciones se aplican solo cuando te conectas al host localhost.

- **`StrictHostKeyChecking no`**: Esta línea establece la política de verificación de clave de host. Cuando StrictHostKeyChecking está configurado en no, SSH no solicitará confirmación antes de agregar automáticamente claves de host desconocidas al archivo known_hosts. Esto puede ser útil para automatizar conexiones a hosts locales donde conoces que la clave de host no cambiará, como en el caso de localhost. Esta configuración especifica que el cliente SSH no realizará comprobaciones estrictas de las claves de host. Esto significa que si la clave de host no se encuentra en el archivo known_hosts, el cliente SSH la aceptará automáticamente sin solicitar al usuario que la confirme. Esto puede ser útil para conexiones a localhost o a direcciones IP locales donde la autenticidad del host es conocida y segura.

Esta configuración es útil en entornos de desarrollo o pruebas donde no necesitas verificar manualmente las claves de host para conexiones locales, pero ten en cuenta que en entornos de producción, es importante mantener la seguridad y mantener activa la verificación de claves de host.
```
Host 0.0.0.0
  StrictHostKeyChecking no
```
Este fragmento del archivo ssh_config establece una configuración específica para las conexiones SSH al host con la dirección IP 0.0.0.0. Veamos qué significa cada parte:

- **`Host 0.0.0.0`**: Esto indica que las siguientes opciones se aplicarán solo cuando te conectes al host con la dirección IP 0.0.0.0. Esta dirección IP se utiliza generalmente como una dirección de enlace a cualquier dirección en una red local.

- **`StrictHostKeyChecking no`**: Aquí se establece la política de verificación de clave de host. Cuando StrictHostKeyChecking se configura en no, SSH no solicitará confirmación antes de agregar automáticamente claves de host desconocidas al archivo known_hosts.

Es importante tener en cuenta que configurar StrictHostKeyChecking en no puede ser riesgoso desde el punto de vista de la seguridad, ya que abre la posibilidad de ataques de intermediarios si la identidad del servidor no está garantizada. Por lo tanto, esta configuración generalmente se usa en situaciones donde la seguridad no es una preocupación, como entornos de prueba o desarrollo local. En producción, es fundamental mantener la verificación de claves de host activa para garantizar la seguridad de las conexiones SSH.
```
Host hadoop-*
   StrictHostKeyChecking no
   UserKnownHostsFile=/dev/null
```
Este fragmento del archivo ssh_config establece una configuración específica para cualquier host cuyo nombre comience con **"hadoop-**". Analicemos lo que significa cada línea:

- **`Host hadoop-*`**: Esto indica que las siguientes opciones se aplicarán a cualquier host cuyo nombre comience con "hadoop-". Por ejemplo, podría aplicarse a hosts con nombres como hadoop-master, hadoop-node1, etc.

- **`StrictHostKeyChecking no`**: Aquí se establece la política de verificación de clave de host. Al configurar StrictHostKeyChecking en no, SSH no solicitará confirmación antes de agregar automáticamente claves de host desconocidas al archivo known_hosts.  Al igual que en las secciones anteriores, esta configuración indica que no se realizarán comprobaciones estrictas de las claves de host para hosts que coincidan con este patrón.

- **`UserKnownHostsFile=/dev/null`**: Esta línea especifica el archivo que SSH utilizará para almacenar las claves de host conocidas. Al establecerlo en /dev/null, se está indicando a SSH que ignore completamente el almacenamiento de claves de host conocidas. Esto significa que SSH no mantendrá ninguna lista de claves de host conocidas, lo que puede ser útil en situaciones donde no es importante la verificación de la identidad del servidor, como entornos de desarrollo o pruebas. Esta configuración especifica que el archivo known_hosts no se utilizará para almacenar las claves de host conocidas. En su lugar, se enviarán todas las claves de host conocidas al dispositivo /dev/null, lo que efectivamente las descarta. Esto significa que el cliente SSH no almacenará las claves de host para los hosts que coincidan con este patrón, lo que puede ser útil en entornos donde las claves de host cambian con frecuencia o donde la autenticidad del host no es crítica.

Es importante tener en cuenta que esta configuración puede ser riesgosa desde el punto de vista de la seguridad, ya que elimina completamente la verificación de la identidad del servidor. Por lo tanto, debe usarse con precaución y solo en entornos donde la seguridad no sea una preocupación, como entornos de desarrollo o pruebas. En entornos de producción, es crucial mantener la verificación de claves de host activa para garantizar la seguridad de las conexiones SSH.

````python
Host localhost
  StrictHostKeyChecking no

Host 0.0.0.0
  StrictHostKeyChecking no
  
Host hadoop-*
   StrictHostKeyChecking no
   UserKnownHostsFile=/dev/null
````
