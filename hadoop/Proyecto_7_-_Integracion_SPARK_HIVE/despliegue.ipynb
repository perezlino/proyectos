##### **Verificamos que la tabla `prescpipeline.sampleTable` pueda ser trabajada en la shell `PYSPARK`**

Esta configuración muestra que la tabla `prescpipeline.sampleTable` es una tabla `EXTERNAL`, lo cual significa que los datos están almacenados fuera de la ubicación predeterminada de Hive. Además, tiene propiedades de almacenamiento específicas y utiliza un formato de entrada y salida determinado por Hive.

Cuando creamos una tabla en Hive y especificas una ubicación (`LOCATION`) diferente a la predeterminada (`/user/hive/warehouse`), esta tabla se convierte automáticamente en una tabla del tipo EXTERNAL. Esto se debe a que:

1. **Tabla administrada (Managed):** Cuando creas una tabla sin especificar una ubicación, Hive se encarga de administrar los datos y los almacena en su directorio predeterminado (`/user/hive/warehouse`). Estas tablas se conocen como tablas administradas o managed tables. Hive asume la responsabilidad de gestionar estos datos, incluyendo la limpieza y el mantenimiento según sea necesario.

2. **Tabla externa (External):** Cuando especificas una ubicación (`LOCATION`) al crear la tabla, estás indicando explícitamente a Hive dónde se encuentran los datos. Hive no asume la responsabilidad de administrar o manejar estos datos, simplemente los registra en su catálogo de metadatos. Estas tablas se conocen como tablas externas o external tables.

Por lo tanto, al especificar una ubicación diferente de la predeterminada (`/user/hive/warehouse`), estás informando a Hive que los datos de la tabla son externos y no deben ser gestionados por Hive. Esto se reflejará en las propiedades de la tabla como tipo EXTERNAL en el catálogo de metadatos de Hive.

Esta configuración muestra que la tabla `db1.dept_partition` es una tabla `EXTERNAL`, lo cual significa que los datos están almacenados fuera de la ubicación predeterminada de Hive. Además, tiene propiedades de almacenamiento específicas y utiliza un formato de entrada y salida determinado por Hive.

Cuando creamos una tabla en Hive y especificas una ubicación (`LOCATION`) diferente a la predeterminada (`/user/hive/warehouse`), esta tabla se convierte automáticamente en una tabla del tipo EXTERNAL. Esto se debe a que:

1. **Tabla administrada (Managed):** Cuando creas una tabla sin especificar una ubicación, Hive se encarga de administrar los datos y los almacena en su directorio predeterminado (`/user/hive/warehouse`). Estas tablas se conocen como tablas administradas o managed tables. Hive asume la responsabilidad de gestionar estos datos, incluyendo la limpieza y el mantenimiento según sea necesario.

2. **Tabla externa (External):** Cuando especificas una ubicación (`LOCATION`) al crear la tabla, estás indicando explícitamente a Hive dónde se encuentran los datos. Hive no asume la responsabilidad de administrar o manejar estos datos, simplemente los registra en su catálogo de metadatos. Estas tablas se conocen como tablas externas o external tables.

Por lo tanto, al especificar una ubicación diferente de la predeterminada (`/user/hive/warehouse`), estás informando a Hive que los datos de la tabla son externos y no deben ser gestionados por Hive. Esto se reflejará en las propiedades de la tabla como tipo EXTERNAL en el catálogo de metadatos de Hive.
