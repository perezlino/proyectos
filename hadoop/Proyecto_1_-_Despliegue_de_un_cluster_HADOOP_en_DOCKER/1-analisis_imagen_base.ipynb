### **Análisis `Dockerfile` para crear `imagen base`**

**`FROM ubuntu:latest`**: Esta línea indica que la imagen que se va a construir se basará en la imagen base de Ubuntu, específicamente la versión "latest" (última). En Docker, una imagen base proporciona el sistema de archivos inicial y los binarios necesarios para ejecutar una aplicación. En este caso, la imagen base es Ubuntu, lo que significa que la imagen de Docker contendrá un sistema de archivos similar al de Ubuntu.

**`MAINTAINER Alfonso Perez Lino`**: Esta línea se utiliza para especificar el nombre del mantenedor o creador de la imagen. En versiones anteriores de Docker, esta línea era importante para proporcionar información sobre quién mantenía o creaba la imagen. Sin embargo, a partir de Docker 1.13, la instrucción **MAINTAINER** está obsoleta en favor de las etiquetas **LABEL**. Ahora, la información del mantenedor se suele agregar como una etiqueta de metadatos en lugar de utilizar la instrucción **MAINTAINER**.

**`LABEL creador="Alfonso Perez Lino"`**: Esta línea agrega metadatos a la imagen resultante. Los metadatos proporcionan información adicional sobre la imagen, como el nombre del creador, la versión, la descripción, etc. En este caso, se está utilizando la etiqueta creador para indicar el nombre del creador de la imagen. Esta información puede ser útil para otros usuarios que trabajen con la imagen, ya que proporciona contexto sobre su origen y su propósito.

```python
# Desplegamos una máquina con el SO Linux Ubuntu
FROM ubuntu:latest

LABEL creador="Alfonso Perez Lino"
```

**`USER root`**: Esta línea indica que se cambiará al usuario root dentro del entorno de la imagen Docker. El usuario root es el superusuario en sistemas basados en Unix y tiene privilegios completos para realizar cualquier acción en el sistema. Al cambiar al usuario root, se obtienen permisos y privilegios completos dentro del entorno de la imagen, lo que permite ejecutar comandos que pueden requerir permisos elevados, como instalar paquetes, configurar el sistema, y realizar otras tareas administrativas.

```python
# Cambiamos al usuario root
USER root
```

**`ENV HADOOP_VERSION 3.3.6`**: Esta línea define una variable de entorno llamada **HADOOP_VERSION** con el valor **3.3.6**. Las variables de entorno se utilizan para almacenar información que puede ser utilizada por scripts, aplicaciones o comandos dentro del entorno de la imagen Docker. En este caso, la variable **HADOOP_VERSION** es utilizada para especificar la versión de Hadoop que se utilizará en la imagen.

**`ENV BASE_DIR /opt/bd`**: Esta línea define otra variable de entorno llamada **BASE_DIR** con el valor **/opt/bd**. Esta variable se utiliza para especificar el directorio base en el cual se instalarán o almacenarán componentes relacionados con Big Data en la imagen Docker.

**`ENV LOG_TAG "[BASE Hadoop_${HADOOP_VERSION}]:"`**: Aquí se define la variable **LOG_TAG** con un valor que parece ser un tag para los logs. Utiliza la variable **HADOOP_VERSION** definida anteriormente para incluir la versión de Hadoop en el tag.

**`ENV REPOSITORY https://dlcdn.apache.org/hadoop/common`**: Por último, esta línea define una variable de entorno llamada **REPOSITORY** con la URL **https://dlcdn.apache.org/hadoop/common**. Esta variable puede utilizarse para especificar la ubicación de un repositorio o recurso externo que la imagen Docker necesita acceder durante su construcción o ejecución. En este caso, corresponde a la URL de un repositorio de distribución de Hadoop.

```python
# Establecemos las variables
ENV HADOOP_VERSION 3.3.6
ENV BASE_DIR /opt/bd
ENV LOG_TAG "[BASE Hadoop_${HADOOP_VERSION}]:"
ENV REPOSITORY https://dlcdn.apache.org/hadoop/common
```

**`RUN echo "$LOG_TAG actualizando e instalando paquetes básicos" && \`**: Esta línea ejecuta un comando que imprimirá un mensaje en la consola indicando que se están actualizando e instalando los paquetes básicos.

**`apt-get update && \`**: Actualiza la lista de paquetes disponibles para su descarga.

**`apt-get install -y --no-install-recommends openjdk-8-jre python3 curl locales iputils-ping && \`**: Instala los paquetes necesarios para Hadoop. Estos paquetes incluyen el entorno de ejecución de Java 8 (openjdk-8-jre), Python 3 (python3), curl (curl), configuración regional (locales) y utilidades de red (iputils-ping). La opción **-y** indica que la instalación debe hacerse sin solicitar confirmación al usuario, y **--no-install-recommends** evita la instalación de paquetes recomendados adicionales.

**`apt-get clean && \`**: Limpia la caché de **apt-get** para reducir el tamaño de la imagen final.

**`rm -rf /var/lib/apt/lists`**: Elimina los archivos de lista de paquetes descargados, lo que también ayuda a reducir el tamaño de la imagen final y elimina información no necesaria.

```python
# Instalamos el software necesario para Hadoop
RUN echo "$LOG_TAG actualizando e instalando paquetes básicos" && \
    apt-get update && \
    apt-get install -y --no-install-recommends openjdk-8-jre python3 curl locales iputils-ping && \
    apt-get clean && \
    rm -rf /var/lib/apt/lists
```

**`RUN echo "$LOG_TAG configurando idioma" && \`**: Esto imprime un mensaje indicando que se está configurando el idioma en el contenedor, utilizando la variable de entorno **$LOG_TAG** para proporcionar una etiqueta en el registro.

**`locale-gen es_ES.UTF-8 &&`**: Este comando genera las configuraciones regionales para el idioma español (**es_ES.UTF-8**). locale-gen es una utilidad de Debian que se utiliza para generar las configuraciones regionales especificadas.

**`update-locale LANG=es_ES.UTF-8`**: Este comando actualiza la configuración regional del sistema para establecer el idioma predeterminado en español (**es_ES.UTF-8**). Establece la variable de entorno **LANG** en **es_ES.UTF-8**, lo que afecta al comportamiento de varios programas y utilidades que dependen de la configuración regional del sistema.

En resumen, esta línea de comandos configura el idioma del sistema en español (`es_ES.UTF-8`) dentro del contenedor Docker durante la creación de la imagen. Esto asegura que las aplicaciones y servicios dentro del contenedor utilicen el idioma español como predeterminado.

```python
# Configuramos el idioma
RUN echo "$LOG_TAG configurando idioma" && \
    locale-gen es_ES.UTF-8 && update-locale LANG=es_ES.UTF-8
```

El comando `RUN mkdir -p ${BASE_DIR}` es una instrucción de Dockerfile utilizada para crear un directorio en la imagen del contenedor. Desglosemos cada parte del comando:

**`RUN`**: Esta palabra clave le dice a Docker que ejecute el comando dentro del contenedor mientras se está construyendo la imagen.

**`mkdir -p`**: Esto es parte del comando **mkdir**, que se utiliza para crear directorios en sistemas Unix. La opción **-p** indica que se deben crear todos los directorios necesarios en la ruta especificada, incluso si los directorios padres aún no existen.

**`${BASE_DIR}`**: Esto es una variable de entorno que se está utilizando para especificar la ruta del directorio que se creará. En el Dockerfile, esta variable debe fue definida anteriormente con una ruta específica, **/opt/bd**.

```python
# Creamos una carpeta para la instalación en /opt
RUN mkdir -p ${BASE_DIR}
```

**`WORKDIR ${BASE_DIR}`**: Este comando establece el directorio de trabajo actual dentro del contenedor. **WORKDIR** se utiliza para cambiar el directorio de trabajo en el contenedor para el resto de los comandos en el Dockerfile. En este caso, se establece el directorio de trabajo como el valor de la variable de entorno **BASE_DIR**, que es **/opt/bd**. Esto significa que todos los comandos posteriores en el Dockerfile se ejecutarán en el contexto de este directorio dentro del contenedor. Esto puede ser útil para organizar el espacio de trabajo y las operaciones del contenedor.

```python
# Cambiamos a la carpeta creada
WORKDIR ${BASE_DIR}
```

El comando **`RUN`** ejecuta una serie de comandos en el contenedor durante la construcción de la imagen:

**`echo "$LOG_TAG Descargando Hadoop"`**: Imprime un mensaje log que utiliza el valor de la variable de entorno LOG_TAG para indicar que se está descargando Hadoop.

**`curl -fLO "${REPOSITORY}/hadoop-${HADOOP_VERSION}/hadoop-${HADOOP_VERSION}.tar.gz"`**: Descarga el archivo tar.gz de la versión especificada de Hadoop desde el repositorio definido en la variable de entorno REPOSITORY.

**`tar xzvf hadoop-${HADOOP_VERSION}.tar.gz`**: Descomprime el archivo tar.gz descargado.

**`ln -s hadoop-${HADOOP_VERSION} hadoop`**: Crea un enlace simbólico llamado hadoop que apunta al directorio recién descomprimido de Hadoop, esto es útil para referenciar la instalación de Hadoop de manera más fácil.

**`rm hadoop-${HADOOP_VERSION}.tar.gz`**: Elimina el archivo tar.gz después de la descompresión para ahorrar espacio en el contenedor.

```python
# Descargamos la última versión de Hadoop, la 3.3.6 en /opt/bd
RUN echo "$LOG_TAG Descargando Hadoop" && \
    curl -fLO "${REPOSITORY}/hadoop-${HADOOP_VERSION}/hadoop-${HADOOP_VERSION}.tar.gz"  && \
    tar xzvf hadoop-${HADOOP_VERSION}.tar.gz && \
    ln -s hadoop-${HADOOP_VERSION} hadoop && \  # Aqui se crea el directorio "hadoop" que corresponde a un "enlace simbolico"
    rm hadoop-${HADOOP_VERSION}.tar.gz
```

Estos son dos comandos de Linux que probablemente se utilicen en un script de inicio o configuración para configurar usuarios y grupos relacionados con Hadoop:

**`groupadd -r hadoop`**: Este comando crea un nuevo grupo llamado "**hadoop**". La opción **`-r`** indica que se debe crear un grupo de sistema, lo que significa que este grupo tendrá un GID (Group Identifier) bajo en el rango de IDs reservados para grupos del sistema.

- **`useradd -r -g hadoop -d ${BASE_DIR} -s /bin/bash hdadmin`**: Este comando crea un nuevo usuario llamado "**hdadmin**". Veamos los detalles de las opciones utilizadas:

    - **`-r`**: Al igual que con **groupadd**, esta opción indica que se debe crear un usuario de sistema, lo que significa que este usuario tendrá un UID (User Identifier) bajo en el rango de IDs reservados para usuarios del sistema. Esta opción indica que se está creando una cuenta de sistema. Las cuentas de sistema generalmente no tienen acceso de inicio de sesión y se usan para ejecutar servicios.

    - **`-g hadoop`**: Esta opción asigna el usuario al grupo "hadoop" que creamos anteriormente. Esto significa que el usuario "**hdadmin**" será miembro del grupo "**hadoop**". Esta opción especifica el grupo principal del nuevo usuario. En este caso, el grupo principal es **hadoop**. Esto significa que el usuario **hdadmin** será añadido al grupo **hadoop**.

    - **`-d ${BASE_DIR}`**: Esta opción especifica el directorio de inicio del usuario. Parece que está utilizando una variable de entorno `${BASE_DIR}` para definir el directorio de inicio. Esto permite una mayor flexibilidad al configurar el directorio de inicio del usuario. Esta opción especifica el directorio de inicio del nuevo usuario. `${BASE_DIR}` es una variable de entorno que debería contener la ruta del directorio de inicio deseado. Por ejemplo, si **BASE_DIR** es **/home/hdadmin**, el directorio de inicio del usuario será **/home/hdadmin**.

    - **`-s /bin/bash`**: Esta opción especifica el shell predeterminado del usuario como "/bin/bash", lo que significa que cuando inicie sesión, se utilizará el shell Bash, se le presentará un entorno Bash.

En resumen, estos comandos están creando un `grupo de sistema` llamado "**hadoop**" y un `usuario de sistema` llamado "**hdadmin**" que pertenece a ese grupo, con su **directorio de inicio** configurado en un directorio definido por la variable de entorno **${BASE_DIR}** y utilizando **Bash** como su shell predeterminado. Esto es típicamente parte de la configuración inicial de un entorno de Hadoop.

```python
# Creamos un grupo Hadoop y un usuario hdadmin para ejecutar los diferentes daemon de Hadoop (HDFS y YARN)
# Cambiamos el propietario del directorio de Hadoop 
RUN groupadd -r hadoop && \
    useradd -r -g hadoop -d ${BASE_DIR} -s /bin/bash hdadmin
```

**`RUN mkdir ${HADOOP_HOME}/logs`**: Ejecuta el comando `mkdir` para crear un directorio llamado `logs` dentro del directorio especificado por la variable de entorno `HADOOP_HOME`.

```python
# Creamos una carpeta para los ficheros de log
RUN mkdir ${HADOOP_HOME}/logs
```

**`RUN`**: Esta es una instrucción de Dockerfile que se utiliza para ejecutar un comando en una nueva capa de la imagen de Docker. Cada instrucción RUN crea una nueva capa en la imagen, y el resultado del comando se guarda en esa capa.

**`chown`**: Este es un comando de Linux que se utiliza para cambiar el propietario de archivos o directorios. "chown" es la abreviatura de "change owner" (cambiar propietario).

**`-R`**: Esta es una opción para el comando chown que significa "recursivamente". Cuando se usa -R, chown cambiará el propietario de todos los archivos y subdirectorios dentro del directorio especificado, así como del directorio en sí.

**`hdadmin:hadoop`**: Esta especificación indica el nuevo propietario y el nuevo grupo del directorio y sus contenidos. Aquí, hdadmin es el usuario y hadoop es el grupo. En otras palabras, después de ejecutar este comando, el usuario hdadmin y el grupo hadoop serán los propietarios del directorio /opt/bd y de todos sus contenidos.

**`${BASE_DIR}`**: Este es el directorio cuyo propietario y grupo se están cambiando.

```python
RUN chown -R hdadmin:hadoop ${BASE_DIR}
```
