#### **Ejecutar contenedor (nodo master): `namenode`**

Vamos a iniciar un contenedor corriendo el servicio **NameNode**:

Este comando de Docker se utiliza para ejecutar un contenedor basado en una imagen llamada `namenode-image`, que está configurada para ser un nodo maestro en un clúster de Hadoop. Aquí está el desglose de cada opción y argumento en el comando:

- **`docker run`**: Es el comando principal de Docker para ejecutar un contenedor.

- **`--rm`**: Esta opción indica a Docker que elimine el contenedor automáticamente después de que se detenga. Esto es útil para limpiar automáticamente los contenedores después de su uso.

- **`--init`**: Esta opción inicia un proceso init dentro del contenedor. Es importante para garantizar que el contenedor pueda manejar correctamente las señales de sistema (como SIGTERM) y realizar una terminación limpia cuando se detiene.

- `--detach (o -d)`: Esta opción ejecuta el contenedor en segundo plano (modo detached), lo que significa que el proceso de Docker se ejecutará en segundo plano y devolverá el control al terminal.

- `--name namenode`: Esta opción establece el nombre del contenedor como namenode. Esto proporciona un nombre identificativo que se puede usar para hacer referencia al contenedor en lugar de usar su ID de contenedor.

- `--network=hadoop-net`: Esta opción especifica la red a la que se unirá el contenedor. En este caso, el contenedor se unirá a una red llamada hadoop-net, que probablemente se haya creado previamente para permitir la comunicación entre los distintos nodos del clúster de Hadoop.

- `--hostname namenode`: Esta opción establece el nombre de host del contenedor como namenode. Esto es útil para identificar el contenedor por su nombre de host dentro de la red.

- `-p 9870:9870`: Esta opción mapea el puerto 9870 del contenedor al puerto 9870 del sistema host. Esto permite acceder a los servicios que se ejecuten en el puerto 9870 dentro del contenedor desde fuera del contenedor, a través del puerto 9870 del sistema host.

- `namenode-image`: Este es el nombre de la imagen Docker que se utilizará para crear el contenedor. Probablemente, esta imagen contiene todas las configuraciones y dependencias necesarias para ejecutar un nodo maestro (NameNode) en un clúster de Hadoop.

En resumen, este comando ejecuta un contenedor Docker basado en la imagen namenode-image, configurado para ser un nodo maestro (NameNode) en un clúster de Hadoop. El contenedor se ejecutará en segundo plano, se unirá a la red hadoop-net, tendrá el nombre de host namenode y expondrá el puerto 9870 para acceder al servicio del NameNode desde fuera del contenedor. Una vez ejecutado, el contenedor se eliminará automáticamente cuando se detenga.

```python
C:\Users\Alfonso\Documents\hadoop-cluster\NameNode>docker run --rm --init --detach --name namenode --network=hadoop-net --hostname namenode -p 9870:9870 namenode-image
e965a9f5191b609c6f5a5653f4cdabc90c30721aee9083c52b7f1f199e50551c

C:\Users\Alfonso\Documents\hadoop-cluster\NameNode>
```

#### **Ejecutar contenedor (nodo master): `resourcemanager`**

Vamos a iniciar un contenedor corriendo el servicio **ResourceManager**:

Este comando de Docker se utiliza para ejecutar un contenedor basado en una imagen llamada `resourcemanager-image`, que está configurada para ser un nodo maestro en un clúster de Hadoop. Aquí está el desglose de cada opción y argumento en el comando:

- **`docker run`**: Es el comando principal de Docker para ejecutar un contenedor.

- **`--rm`**: Esta opción indica a Docker que elimine el contenedor automáticamente después de que se detenga. Esto es útil para limpiar automáticamente los contenedores después de su uso.

- **`--init`**: Esta opción inicia un proceso init dentro del contenedor. Es importante para garantizar que el contenedor pueda manejar correctamente las señales de sistema (como SIGTERM) y realizar una terminación limpia cuando se detiene.

- `--detach (o -d)`: Esta opción ejecuta el contenedor en segundo plano (modo detached), lo que significa que el proceso de Docker se ejecutará en segundo plano y devolverá el control al terminal.

- `--name resourcemanager`: Esta opción establece el nombre del contenedor como resourcemanager. Esto proporciona un nombre identificativo que se puede usar para hacer referencia al contenedor en lugar de usar su ID de contenedor.

- `--network=hadoop-net`: Esta opción especifica la red a la que se unirá el contenedor. En este caso, el contenedor se unirá a una red llamada hadoop-net, que probablemente se haya creado previamente para permitir la comunicación entre los distintos nodos del clúster de Hadoop.

- `--hostname resourcemanager`: Esta opción establece el nombre de host del contenedor como resourcemanager. Esto es útil para identificar el contenedor por su nombre de host dentro de la red.

- `-p 8088:8088`: Esta opción mapea el puerto 8088 del contenedor al puerto 8088 del sistema host. Esto permite acceder a los servicios que se ejecuten en el puerto 8088 dentro del contenedor desde fuera del contenedor, a través del puerto 8088 del sistema host.

- `resourcemanager-image`: Este es el nombre de la imagen Docker que se utilizará para crear el contenedor. Probablemente, esta imagen contiene todas las configuraciones y dependencias necesarias para ejecutar un nodo maestro (ResourceManager) en un clúster de Hadoop.

En resumen, este comando ejecuta un contenedor Docker basado en la imagen resourcemanager-image, configurado para ser un nodo maestro (ResourceManager) en un clúster de Hadoop. El contenedor se ejecutará en segundo plano, se unirá a la red hadoop-net, tendrá el nombre de host resourcemanager y expondrá el puerto 8088 para acceder al servicio del ResourceManager desde fuera del contenedor. Una vez ejecutado, el contenedor se eliminará automáticamente cuando se detenga.

```python
C:\Users\Alfonso\Documents\hadoop-cluster\ResourceManager>docker run --rm --init --detach --name resourcemanager --network=hadoop-net --hostname resourcemanager -p 8088:8088 resourcemanager-image
3e84a3ece33a7335815abaafb733bc7c41425f57385d29cb61d21781385e9cb7

C:\Users\Alfonso\Documents\hadoop-cluster\ResourceManager>
```

Este comando de Docker se utiliza para ejecutar un contenedor basado en una imagen llamada `dnnm-image`, que está configurada para ser un nodo maestro en un clúster de Hadoop. Aquí está el desglose de cada opción y argumento en el comando:

- **`docker run`**: Es el comando principal de Docker para ejecutar un contenedor.

- **`--rm`**: Esta opción indica a Docker que elimine el contenedor automáticamente después de que se detenga. Esto es útil para limpiar automáticamente los contenedores después de su uso.

- **`--init`**: Esta opción inicia un proceso init dentro del contenedor. Es importante para garantizar que el contenedor pueda manejar correctamente las señales de sistema (como SIGTERM) y realizar una terminación limpia cuando se detiene.

- `--detach (o -d)`: Esta opción ejecuta el contenedor en segundo plano (modo detached), lo que significa que el proceso de Docker se ejecutará en segundo plano y devolverá el control al terminal.

- `--name dnnm1`: Esta opción establece el nombre del contenedor como dnnm1. Esto proporciona un nombre identificativo que se puede usar para hacer referencia al contenedor en lugar de usar su ID de contenedor.

- `--network=hadoop-net`: Esta opción especifica la red a la que se unirá el contenedor. En este caso, el contenedor se unirá a una red llamada hadoop-net, que probablemente se haya creado previamente para permitir la comunicación entre los distintos nodos del clúster de Hadoop.

- `--hostname dnnm1`: Esta opción establece el nombre de host del contenedor como dnnm1. Esto es útil para identificar el contenedor por su nombre de host dentro de la red.

- `dnnm-image`: Este es el nombre de la imagen Docker que se utilizará para crear el contenedor. Probablemente, esta imagen contiene todas las configuraciones y dependencias necesarias para ejecutar un nodo worker (DataNode-NodeManager) en un clúster de Hadoop.

En resumen, este comando ejecuta un contenedor Docker basado en la imagen dnnm-image, configurado para ser un nodo worker (DataNode-NodeManager) en un clúster de Hadoop. El contenedor se ejecutará en segundo plano, se unirá a la red hadoop-net, tendrá el nombre de host dnnm1. Una vez ejecutado, el contenedor se eliminará automáticamente cuando se detenga.

#### **`Práctica`: crear un archivo y moverlo a `HDFS` (Visualizar en cuantos bloques fue dividido ese archivo y donde se alamcenan)**

**`dd if=/dev/urandom of=fich300M bs=1M count=300`**: esta línea de comando en Linux utiliza el comando `dd` para generar un archivo de **300MB** lleno de datos aleatorios. Aquí está el desglose:

- **`dd`**: Es un comando de Linux utilizado para copiar y convertir archivos. Puede ser peligroso si se utiliza incorrectamente, ya que puede sobrescribir datos fácilmente.

- **`if=/dev/urandom`**: Esta parte de la línea de comando especifica la fuente de los datos que se utilizarán para llenar el archivo. /dev/urandom es un dispositivo especial en Unix/Linux que proporciona una secuencia de bytes aleatorios.

- **`of=fich300M`**: Esta parte especifica el nombre del archivo de salida que se creará. En este caso, el archivo se llamará "fich300M". Si no existe, se creará; si ya existe, se sobrescribirá.

- **`bs=1M`**: Esto establece el tamaño del bloque de entrada y salida en 1 megabyte (1M). Especifica cuántos bytes se leen y escriben a la vez.

- **`count=300`**: Esta opción establece el número de bloques de entrada que se copiarán. Dado que el tamaño de bloque se ha establecido en 1MB y el recuento en 300, el archivo resultante tendrá un tamaño de 300MB (300 bloques * 1MB/bloque = 300MB).

Entonces, en resumen, la línea de comando está utilizando el comando dd para tomar datos aleatorios del dispositivo /dev/urandom y escribirlos en un archivo llamado "fich300M", con un tamaño de 300MB, utilizando bloques de 1MB cada uno.

```python
C:\Users\Alfonso>docker exec -it namenode bash
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ dd if=/dev/urandom of=fich300M bs=1M count=300
300+0 records in
300+0 records out
314572800 bytes (315 MB, 300 MiB) copied, 2.93992 s, 107 MB/s
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ ls
LICENSE-binary  LICENSE.txt  NOTICE-binary  NOTICE.txt  README.txt  bin  etc  fich300M  include  lib  libexec  licenses-binary  logs  sbin  share  test.txt
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ hdfs dfs -put fich300M /test  # <----- Copiamos el archivo "fich300M" a HDFS en el directorio /test
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ rm fich300M  # <----- Borramos el archivo
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ LS
bash: LS: command not found
hdadmin@namenode:~/hadoop-3.3.6$ ls
LICENSE-binary  LICENSE.txt  NOTICE-binary  NOTICE.txt  README.txt  bin  etc  include  lib  libexec  licenses-binary  logs  sbin  share  test.txt
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$ hdfs dfs -ls /test  # <----- Buscamos el archivo en HDFS
Found 1 items
-rw-r--r--   3 hdadmin supergroup  314572800 2024-05-20 22:52 /test/fich300M  # <----- El archivo tiene un numero 3. Dice que fue replicado 3 veces
hdadmin@namenode:~/hadoop-3.3.6$
hdadmin@namenode:~/hadoop-3.3.6$
```
