### **Extraer los datos de fechas, partidos y goles de todos los equipos de la `Premier League`, visualizarlos en un Dataframe y exportarlos a un archivo CSV**

Usamos el giguiente sitio web:

https://www.adamchoi.co.uk/overs/detailed

Vamos a extraer los datos de fechas, partidos y goles de todos los equipos de la Premier League. El sitio se encuentra con el boton de `Home/Away split` activado por defecto el cual nos permite ver los goles de cada equipo en tablas separadas para los que son partidos de local y partidos de visita:

<center><img src="https://i.postimg.cc/Df9390CC/ws-294.png"></center>

Inspeccionamos el sitio y buscamos el tag que almacene el boton **`All matches`**:

<center><img src="https://i.postimg.cc/9QS2W6wD/ws-4.png"></center>
<center><img src="https://i.postimg.cc/59SdTYxK/ws-6.png"></center>

Con este código logramos pulsar el botón de `All matches` el cual nos redirige a otra pagina de la cual extraremos data que se encuentra en tablas:

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

website = 'https://www.adamchoi.co.uk/overs/detailed'
# Se recomienda utilizar Chrome, pero podriamos utilizar Firefox, Safari, Edge, etc.
driver = webdriver.Chrome()
driver.get(website)

all_matches_button = driver.find_element(by=By.XPATH, value='//label[@analytics-event="All matches"]')
all_matches_button.click()

# driver.quit() # Nos sirve para cerrar la ventana que se nos abre
```

Luego de generar el click en el boton nos envia a la siguiente página, donde los goles de local y visita se ordenan en una sola tabla:

<center><img src="https://i.postimg.cc/65b54tzn/ws-293.png"></center>

Así que normalmente cuando tengas que extraer datos de tablas, verás que las tablas utilizan TagNames que son `<tr>` o `<td>`. Así, por ejemplo, `<tr>` representa losd atos de cada fila y `<td>` va a representar los datos en cada columna. 

En el punto anterior, vimos como pulsar un botón desde código. Vamos a comenzar desde ahi:

<center><img src="https://i.postimg.cc/6qHX6t8T/ws-7.png"></center>

Existen 606 filas con datos de partidos. Existen 606 tag `<tr>` que poseen la data que necesitamos extraer:

<center><img src="https://i.postimg.cc/CLJdC9HH/ws-295.png"></center>

Este tag `<tr>` representa a la primera fila de la primera tabla con datos del Arsenal. Cada equipo de la Premier League tiene su propia tabla con datos:

<center><img src="https://i.postimg.cc/9FzHSSHB/ws-296.png"></center>

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

website = 'https://www.adamchoi.co.uk/overs/detailed'
# Se recomienda utilizar Chrome, pero podriamos utilizar Firefox, Safari, Edge, etc.
driver = webdriver.Chrome()
driver.get(website)

all_matches_button = driver.find_element(by=By.XPATH, value='//label[@analytics-event="All matches"]')
all_matches_button.click()

partidos = driver.find_elements(by=By.XPATH, value='//tr[@data-ng-repeat="match in ::dtc.$scope.matches"]')

fecha = []
equipo_local = []
resultado = []
equipo_visita = []

# Así que como vemos aquí, ya no estamos usando "driver.find_element(by=By.XPATH)", porque queremos usar el contexto "partido". 
# Así que queremos tener "partidos" como referencia para nuestro XPath. Así que si queremos usar ese contexto "partido", tenemos que 
# escribir "partido" en lugar de driver. Y luego en el XPath, tenemos que escribir el signo de "punto", porque ese signo de punto 
# especifica que queremos usar el contexto actual y luego tenemos que construir ese XPath.

for partido in partidos:
    fecha.append(partido.find_element(by=By.XPATH, value='./td[1]').text)
    equipo_local.append(partido.find_element(by=By.XPATH, value='./td[2]').text)
    resultado.append(partido.find_element(by=By.XPATH, value='./td[3]').text)
    equipo_visita.append(partido.find_element(by=By.XPATH, value='./td[4]').text)

# driver.quit() # Nos sirve para cerrar la ventana que se nos abre
```

Ahora vamos a exportar la data extraida en un archivo **`CSV`**:

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

website = 'https://www.adamchoi.co.uk/overs/detailed'
# Se recomienda utilizar Chrome, pero podriamos utilizar Firefox, Safari, Edge, etc.
driver = webdriver.Chrome()
driver.get(website)

all_matches_button = driver.find_element(by=By.XPATH, value='//label[@analytics-event="All matches"]')
all_matches_button.click()

partidos = driver.find_elements(by=By.XPATH, value='//tr[@data-ng-repeat="match in ::dtc.$scope.matches"]')

fecha = []
equipo_local = []
resultado = []
equipo_visita = []

# Así que como vemos aquí, ya no estamos usando "driver.find_element(by=By.XPATH)", porque queremos usar el contexto "partido". 
# Así que queremos tener "partidos" como referencia para nuestro XPath. Así que si queremos usar ese contexto "partido", tenemos que 
# escribir "partido" en lugar de driver. Y luego en el XPath, tenemos que escribir el signo de "punto", porque ese signo de punto 
# especifica que queremos usar el contexto actual y luego tenemos que construir ese XPath.

for partido in partidos:
    fecha.append(partido.find_element(by=By.XPATH, value='./td[1]').text)
    equipo_local.append(partido.find_element(by=By.XPATH, value='./td[2]').text)
    resultado.append(partido.find_element(by=By.XPATH, value='./td[3]').text)
    equipo_visita.append(partido.find_element(by=By.XPATH, value='./td[4]').text)

driver.quit() # Nos sirve para cerrar la ventana que se nos abre
    
# Crear Dataframe en Pandas y exportar a CSV (Excel)
df = pd.DataFrame({'fecha': fecha, 'equipo_local': equipo_local, 'resultado': resultado, 'equipo_visita': equipo_visita})
df.to_csv('football_data.csv', index=False)
df
```
