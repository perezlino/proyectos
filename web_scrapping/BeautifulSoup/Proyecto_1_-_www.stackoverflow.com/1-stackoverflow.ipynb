Inspeccionamos el sitio y buscamos la caja que almacene las preguntas:

Llamamos a todos los tag **`div`** que contengan la clase **`s-post-summary--content`** directamente en la variable **`lista_de_preguntas`** que se encuentran en la caja **`contenedor_de_preguntas`**. En este div se encuentran tanto las **preguntas**, como la **descripcion de las preguntas**, por tanto, nos facilita para obtener ambas:

```python
import requests
from bs4 import BeautifulSoup

# USER AGENT PARA PROTEGERNOS DE BANEOS
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# URL SEMILLA
url = 'https://stackoverflow.com/questions'

# REQUERIMIENTO AL SERVIDOR
respuesta = requests.get(url, headers=headers)

# PARSEO DEL ARBOL CON BEAUTIFUL SOUP
soup = BeautifulSoup(respuesta.text)

contenedor_de_preguntas = soup.find(id="questions") # ENCONTRAR UN ELEMENTO POR ID
lista_de_preguntas = contenedor_de_preguntas.find_all('div', class_="s-post-summary--content") # ENCONTRAR VARIOS ELEMENTOS POR TAG Y POR CLASE
```

```python
len(lista_de_preguntas)
```

Tengo que buscar el tag **`h3`** para encontrar la pregunta:

```python
for pregunta in lista_de_preguntas:
    texto_pregunta = pregunta.find('h3').get_text(strip=True) # DENTRO DE CADA ELEMENTO ITERADO ENCONTRAR UN TAG
    print(texto_pregunta)
```

Ahora buscamos tanto la pregunta que se encuentra en el tag **`h3`** y su descripcion que se encuentra en el **`div`** con la clase **`s-post-summary--content-excerpt`**:

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO

  texto_pregunta = pregunta.find('h3').get_text(strip=True) # DENTRO DE CADA ELEMENTO ITERADO ENCONTRAR UN TAG
  descripcion_pregunta = pregunta.find(class_='s-post-summary--content-excerpt').text # ENCONTRAR POR CLASE
  print (f'{texto_pregunta}:{descripcion_pregunta}')
```

Podemos limpiar un poco lo devuelto:

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO

  texto_pregunta = pregunta.find('h3').get_text(strip=True) # DENTRO DE CADA ELEMENTO ITERADO ENCONTRAR UN TAG
  descripcion_pregunta = pregunta.find(class_='s-post-summary--content-excerpt').text # ENCONTRAR POR CLASE
  descripcion_pregunta = descripcion_pregunta.replace('\n', '') # LIMPIEZA DE TEXTO
  print (f'{texto_pregunta}:{descripcion_pregunta}')
```

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO

  texto_pregunta = pregunta.find('h3').get_text(strip=True) # DENTRO DE CADA ELEMENTO ITERADO ENCONTRAR UN TAG
  descripcion_pregunta = pregunta.find(class_='s-post-summary--content-excerpt').get_text(strip=True) # ENCONTRAR POR CLASE
  descripcion_pregunta = descripcion_pregunta.replace('\n', '') # LIMPIEZA DE TEXTO
  print (f'{texto_pregunta}:{descripcion_pregunta}')
```

Llamamos a todos los tag **`div`** que contengan la clase **`s-post-summary--content`** directamente en la variable **`lista_de_preguntas`** que se encuentran en la caja **`contenedor_de_preguntas`**. En este div se encuentran tanto las **preguntas**, como la **descripcion de las preguntas**, por tanto, nos facilita para obtener ambas:

Ambas se encuentran al MISMO NIVEL:

```python
import requests
from bs4 import BeautifulSoup

# USER AGENT PARA PROTEGERNOS DE BANEOS
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# URL SEMILLA
url = 'https://stackoverflow.com/questions'

# REQUERIMIENTO AL SERVIDOR
respuesta = requests.get(url, headers=headers)

# PARSEO DEL ARBOL CON BEAUTIFUL SOUP
soup = BeautifulSoup(respuesta.text)

contenedor_de_preguntas = soup.find(id="questions") # ENCONTRAR UN ELEMENTO POR ID
lista_de_preguntas = contenedor_de_preguntas.find_all('div', class_="s-post-summary--content") # ENCONTRAR VARIOS ELEMENTOS POR TAG Y POR CLASE
```

```python
len(lista_de_preguntas)
```

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO  
  contenedor_pregunta = pregunta.find('h3')
  texto_pregunta = contenedor_pregunta.get_text(strip=True)
  print(texto_pregunta)
```

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO  
  contenedor_pregunta = pregunta.find('h3')
  texto_pregunta = contenedor_pregunta.get_text(strip=True)
  descripcion_pregunta = contenedor_pregunta.find_next_sibling('div') # NOS TRAE EL ELEMENTO HERMANO (SE ENCUENTRA EN EL MISMO NIVEL)
  print(descripcion_pregunta)
```

```python
for pregunta in lista_de_preguntas: # ITERAR ELEMENTO POR ELEMENTO  
  contenedor_pregunta = pregunta.find('h3')
  texto_pregunta = contenedor_pregunta.get_text(strip=True)
  descripcion_pregunta = contenedor_pregunta.find_next_sibling('div') # NOS TRAE EL ELEMENTO HERMANO (SE ENCUENTRA EN EL MISMO NIVEL)
  texto_descripcion_pregunta = descripcion_pregunta.get_text(strip=True)
  texto_descripcion_pregunta = texto_descripcion_pregunta.replace('\n', '')
  print(texto_descripcion_pregunta)
```
