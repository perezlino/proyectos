#### **Código**

Recordar que el método `parse()` será llamado para manejar cada una de las requests para esas URLs, aunque no le hayamos dicho explícitamente a Scrapy que lo haga. Esto sucede porque `parse()` es el método `callback` por defecto de Scrapy, que es llamado para las requests sin un callback explícitamente asignado.

```python
import scrapy


class QuotesToScrapeCuatroSpider(scrapy.Spider):
    name = "quotes_to_scrape_cuatro"
    start_urls = [
        "https://quotes.toscrape.com/page/1/",
        "https://quotes.toscrape.com/page/2/",
    ]

    # Trabajando con selectores CSS
    # def parse(self, response):
    #     for quote in response.css("div.quote"):
    #         yield {
    #             "text": quote.css("span.text::text").get(),
    #             "author": quote.css("small.author::text").get(),
    #             "tags": quote.css("div.tags a.tag::text").getall(),
    #            }

    def parse(self, response):
        # Cada vez que queramos localizar un elemento, vamos a utilizar la variable 'response'
        for quote in response.xpath('//div[@class="quote"]'):
            yield {
                # El método 'extract_first() aún se utiliza, pero ahora se recomienda usar get() ---> Extrae 1 elemento
                'text': quote.xpath('./span[@class="text"]/text()').extract_first(),
                'author': quote.xpath('.//small[@class="author"]/text()').extract_first(),
                # El método 'extract() aún se utiliza, pero ahora se recomienda usar getall() ---> Extrae varios elementos --> Lista
                'tags': quote.xpath('.//div[@class="tags"]/a[@class="tag"]/text()').extract()
            }
```

#### **Ejecución**

Nos posicionamos en la ruta correcta y ejecutamos nuestra Spider:

```python
scrapy crawl quotes_to_scrape_cuatro
```

Esto generará un archivo `quotes_to_scrape_cuatro.json` que contendrá todos los elementos scrapeados, serializados en JSON.

El modificador de línea de comandos `-O` sobrescribe cualquier archivo existente; utilice `-o` en su lugar para añadir nuevo contenido a cualquier archivo existente. Sin embargo, al añadir contenido a un archivo JSON, el contenido del archivo se convierte en JSON no válido. Cuando añada contenido a un archivo, considere utilizar un formato de serialización diferente, como JSON Lines:

**`scrapy crawl quotes_to_scrape -o quotes_to_scrape_cuatro.jsonl`**
