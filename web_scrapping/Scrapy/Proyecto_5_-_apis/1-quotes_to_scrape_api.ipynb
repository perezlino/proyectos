#### **Código**

```python
import scrapy
import json

class QuotesToScrapeApiSpider(scrapy.Spider):
    name = "quotes_to_scrape_api"
    allowed_domains = ["quotes.toscrape.com"]
    start_urls = ["https://quotes.toscrape.com/api/quotes?page=1"]

    def parse(self, response):
        # Almacenar la respuesta en json y obtener quotes
        json_response = json.loads(response.body)
        quotes = json_response.get('quotes')

        # Recorrer los elementos de quotes
        for quote in quotes:
            # Devuelve los datos extraídos
            yield {
                'autor': quote.get('author').get('name'),
                'tags': quote.get('tags'),
                'cita': quote.get('text'),
            }
```

#### **Ejecución**

Nos posicionamos en la ruta correcta y ejecutamos nuestra Spider:

```python
scrapy crawl quotes_to_scrape_api
```
