##### **`Forma 1`**

```python
import scrapy


class WorldmetersTresSpider(scrapy.Spider):
    name = "worldmeters_tres"
    allowed_domains = ["www.worldometers.info"]
    start_urls = ["https://www.worldometers.info/world-population/population-by-country"]

    def parse(self, response):
        # Extracción de nombres de paises y su población
        filas = response.xpath('//tbody/tr')

        for fila in filas:
            pais = fila.xpath('./td/a/text()').get()
            # Obtenemos un 'Enlace relativo' --> /world-population/india-population/
            enlace = fila.xpath('./td/a/@href').get()

            # Devuelve los datos extraidos
            # Vamos a devolver un 'Enlace absoluto'
            yield {
                'pais':pais,
                'enlace': f'https://www.worldometers.info{enlace}' # https://www.worldometers.info/world-population/india-population/
            }
```

##### **`Forma 2`**

```python
import scrapy


class WorldmetersTresSpider(scrapy.Spider):
    name = "worldmeters_tres"
    allowed_domains = ["www.worldometers.info"]
    start_urls = ["https://www.worldometers.info/world-population/population-by-country"]

    def parse(self, response):
        # Extracción de nombres de paises y su población
        filas = response.xpath('//tbody/tr')

        for fila in filas:
            pais = fila.xpath('./td/a/text()').get()
            # Obtenemos un 'Enlace relativo' --> /world-population/india-population/
            enlace = fila.xpath('./td/a/@href').get()

            # Devuelve los datos extraidos
            # Vamos a devolver un 'Enlace absoluto'
            yield {
                'pais':pais,
                'enlace': response.urljoin(enlace) # https://www.worldometers.info/world-population/india-population/
            }
```
