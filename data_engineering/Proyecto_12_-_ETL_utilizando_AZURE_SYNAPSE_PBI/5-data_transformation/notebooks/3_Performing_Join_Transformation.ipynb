```python
# Lectura del dataframe para la operación de join

df_tojoin = spark.read.format('csv')\
                        .option('header','true')\
                        .load('abfss://refined@adlsproyectos.dfs.core.windows.net/ForJoins/Education_and_expected_salary_ranges.csv')
```

```python
# Lectura de los datos transformados de la transformación anterior

df = spark.read.format('parquet')\
                        .load('abfss://refined@adlsproyectos.dfs.core.windows.net/DataTransformed/*.parquet')
```

```python
# Aplicamos la transformación join 

df_joined = df.join(df_tojoin,on='Education_Level',how='inner')
```

```python
# Visualizamos el dataframe resultado

display(df_joined)
```

```python
# Escritura de los datos transformados en el Data Lake

df_joined.write.format('parquet')\
                .mode('overwrite')\
                .save('abfss://refined@adlsproyectos.dfs.core.windows.net/JoinedData/')
```
