```python
from pyspark.sql.window import Window
from pyspark.sql.functions import dense_rank
```

```python
# Lectura de los datos transformados de la transformación anterior

df = spark.read.format('parquet')\
                        .load('abfss://refined@adlsproyectos.dfs.core.windows.net/StringSorted/*.parquet')
```

```python
# Definir una ventana que ordena los datos en función de la tasa de desempleo (UnEmployed_Rate_Percentage)

window = Window.orderBy('UnEmployed_Rate_Percentage')
```

```python
# Obtener el ranking basado en la tasa de desempleo (UnEmployed_Rate_Percentage) utilizando Dense Rank

df_denserank = df.withColumn('dense_rank',dense_rank().over(window))
```

```python
# Visualizamos el dataframe resultado

display(df_denserank)
```

```python
# Escritura del dataframe transformado en Data Lake

df_denserank.write.format('parquet')\
                .mode('overwrite')\
                .save('abfss://refined@adlsproyectos.dfs.core.windows.net/WindowFunctions/')
```
