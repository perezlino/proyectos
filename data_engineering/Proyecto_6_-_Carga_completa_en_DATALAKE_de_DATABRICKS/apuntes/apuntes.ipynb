### **Apuntes #3** - Carga Incremental para el `MÃ©todo 3` (Parte 2)

Esta funciÃ³n nos genera una lista con el orden necesario de las columnas, dejando al final la columna **race_id**

1. **input_df** es el parÃ¡metro que ingresamos que corresponde al dataframe
2. **partition_column** es el parÃ¡metro que debemos ingresar igual a **race_id**
3. **input_df.schema.names** nos devuelve una lista con los nombres de las columnas del dataframe **input_df**

````python
# La funciÃ³n "re_arrange_partition_column()" se encuentra en el notebook "common_functions"
# Recordar que esta funciÃ³n se llamarÃ¡ al utilizar %run "../includes/common_functions"
def re_arrange_partition_column(input_df, partition_column):
  column_list = []
  for column_name in input_df.schema.names:
    if column_name != partition_column: # Si "column_name" es distinto a "race_id"
      column_list.append(column_name)  # Lo aÃ±adirÃ¡ a la lista "column_list"
  column_list.append(partition_column) # Despues de ejecutarse el FOR, agrega a la lista "column_list" el valor de "race_id" al Ãºltimo
  output_df = input_df.select(column_list)
  return output_df
````

````python
re_arrange_partition_column(results_final_df, 'race_id')
````

1. Esta funciÃ³n llama a la funciÃ³n **re_arrange_partition_column** de la cual obtiene el dataframe con el orden que necesitamos de las columnas, con la columna particionada **race_id** al Ãºltimo.
2. Finalmente se ejecuta la condiciÃ³n **IF** y si existiese la tabla, se sobrescribirÃ¡n las particiones coincidentes
3. En caso contrario, se crearÃ¡ la tabla

````python
# La funciÃ³n "overwrite_partition()" se encuentra en el notebook "common_functions"
# Recordar que esta funciÃ³n se llamarÃ¡ al utilizar %run "../includes/common_functions"
def overwrite_partition(input_df, db_name, table_name, partition_column):
  output_df = re_arrange_partition_column(input_df, partition_column)
  spark.conf.set("spark.sql.sources.partitionOverwriteMode","dynamic")
  if (spark._jsparkSession.catalog().tableExists(f"{db_name}.{table_name}")):
    output_df.write.mode("overwrite").insertInto(f"{db_name}.{table_name}")
  else:
    output_df.write.mode("overwrite").partitionBy(partition_column).format("parquet").saveAsTable(f"{db_name}.{table_name}")
````

#### Escribir datos en el contenedor **presentation** del ADLS como **parquet** y crear la tabla **constructor_standings** en la base de datos **f1_presentation**

#### **Vamos a analizar las funciones**

Esta funciÃ³n nos genera una lista con el orden necesario de las columnas, dejando al final la columna **race_year**

1. **input_df** es el parÃ¡metro que ingresamos que corresponde al dataframe
2. **partition_column** es el parÃ¡metro que debemos ingresar igual a **race_year**
3. **input_df.schema.names** nos devuelve una lista con los nombres de las columnas del dataframe **input_df**

````python
# La funciÃ³n "re_arrange_partition_column()" se encuentra en el notebook "common_functions"
# Recordar que esta funciÃ³n se llamarÃ¡ al utilizar %run "../includes/common_functions"
def re_arrange_partition_column(input_df, partition_column):
  column_list = []
  for column_name in input_df.schema.names:
    if column_name != partition_column: # Si "column_name" es distinto a "race_year"
      column_list.append(column_name)  # Lo aÃ±adirÃ¡ a la lista "column_list"
  column_list.append(partition_column) # Despues de ejecutarse el FOR, agrega a la lista "column_list" el valor de "race_year" al Ãºltimo
  output_df = input_df.select(column_list)
  return output_df
````

````python
re_arrange_partition_column(final_df, 'race_year')
````

1. Esta funciÃ³n llama a la funciÃ³n **re_arrange_partition_column** de la cual obtiene el dataframe con el orden que necesitamos de las columnas, con la columna particionada **race_year** al Ãºltimo.
2. Finalmente se ejecuta la condiciÃ³n **IF** y si existiese la tabla, se sobrescribirÃ¡n las particiones coincidentes
3. En caso contrario, se crearÃ¡ la tabla

````python
# La funciÃ³n "overwrite_partition()" se encuentra en el notebook "common_functions"
# Recordar que esta funciÃ³n se llamarÃ¡ al utilizar %run "../includes/common_functions"
def overwrite_partition(input_df, db_name, table_name, partition_column):
  output_df = re_arrange_partition_column(input_df, partition_column)
  spark.conf.set("spark.sql.sources.partitionOverwriteMode","dynamic")
  if (spark._jsparkSession.catalog().tableExists(f"{db_name}.{table_name}")):
    output_df.write.mode("overwrite").insertInto(f"{db_name}.{table_name}")
  else:
    output_df.write.mode("overwrite").partitionBy(partition_column).format("parquet").saveAsTable(f"{db_name}.{table_name}")
````

Por Ãºltimo, llamamos a la funciÃ³n:

````python
# La funciÃ³n "overwrite_partition()" se encuentra en el notebook "common_functions"
# Recordar que esta funciÃ³n se llamarÃ¡ al utilizar %run "../includes/common_functions"
overwrite_partition(final_df, 'f1_presentation', 'constructor_standings', 'race_year')
````
