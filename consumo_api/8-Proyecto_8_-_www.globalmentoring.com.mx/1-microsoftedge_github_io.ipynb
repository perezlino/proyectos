### **Obtener información de la API generada por el sitio `globalmentoring.com.mx/api`**

#### **`Enlace 1`**

##### **Conexión a la API**

- El problema, es que algunos servidores necesitan la cabecera **`User-Agent`** y este parece ser uno de ellos
  (https://developer.mozilla.org/es/docs/Web/HTTP/Headers/User-Agent)
- Por defecto, los navegadores ya añaden esa cabecera a la petición
- Para solucionarlo, podemos preparar la petición usando la clase **`Request`** y añadir esa cabecera

```python
import requests
import json
import pandas as pd

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'}

response = requests.get("http://globalmentoring.com.mx/api/personas.json", headers=headers)
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

Podemos usar tambien la libreria `urllib`:

```python
import json
from urllib.request import Request
from urllib.request import urlopen
import pandas as pd
from pprint import pprint

request = Request("http://globalmentoring.com.mx/api/personas.json")

# Añadir la cebecera User-Agent a la peticion
request.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59')
respuesta  = urlopen(request)
```

##### **Extraer información en el tipo `diccionario`**

```python
cuerpo_respuesta = respuesta.read()

json_respuesta = json.loads(cuerpo_respuesta.decode('utf-8'))
pprint(json_respuesta)
print(type(json_respuesta)) # <class 'dict'>
```

- Obtener información diversa desde la API:

```python
for persona in json_respuesta['personas']:
    print(persona, f'// {type(persona)}')
```

```python
for persona in json_respuesta['personas']:
    print(f'Persona: {persona["nombre"]}, Edad: {persona["edad"]}')
```

```python
# Accedemos a las variables independientes
print(f'Total de personas: {json_respuesta["total"]}')
print(f'Mensaje: {json_respuesta["mensaje"]}')
```

#### **`Enlace 2`**

##### **Conexión a la API**

```python
import requests
import json
import pandas as pd

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'}

response = requests.get("http://globalmentoring.com.mx/api/clima.json", headers=headers)
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

Podemos usar tambien la libreria `urllib`:

```python
import json
from urllib.request import Request
from urllib.request import urlopen
import pandas as pd
from pprint import pprint

request = Request("http://globalmentoring.com.mx/api/clima.json")

# Añadir la cebecera User-Agent a la peticion
request.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59')
respuesta  = urlopen(request)
```

##### **Extraer información en el tipo `diccionario`**

```python
cuerpo_respuesta = respuesta.read()

json_respuesta = json.loads(cuerpo_respuesta.decode('utf-8'))
pprint(json_respuesta)
print(type(json_respuesta)) # <class 'dict'>
```

JSON se convierte a listas y diccionarios en Python

```python
descripcion_clima = json_respuesta.get('clima')
print(descripcion_clima)
# [{'principal': 'Con nubes', 'descripcion': 'Pocas nubes'}]

descripcion_clima = json_respuesta.get('clima')[0]
print(descripcion_clima)
# {'principal': 'Con nubes', 'descripcion': 'Pocas nubes'}

descripcion_clima = json_respuesta.get('clima')[0].get('descripcion')
print(f'Descripción clima: {descripcion_clima}')
# Descripción clima: Pocas nubes

descripcion_clima = json_respuesta['clima'][0]['descripcion']
print(f'Descripción clima: {descripcion_clima}')
# Descripción clima: Pocas nubes
```

```python
for clima in json_respuesta['clima']:
    print(clima)
# {'principal': 'Con nubes', 'descripcion': 'Pocas nubes'}

for clima in json_respuesta['clima']:
    print(f'Descripción clima: {clima["descripcion"]}')
# Descripción clima: Pocas nubes
```

```python
temp_min = json_respuesta['principal']['temp_min']
print(f'Temperaratura mínima: {temp_min}')
# Temperaratura mínima: 11

temp_min = json_respuesta.get('principal').get('temp_min')
print(f'Temperaratura mínima: {temp_min}')
# Temperaratura mínima: 11
```
