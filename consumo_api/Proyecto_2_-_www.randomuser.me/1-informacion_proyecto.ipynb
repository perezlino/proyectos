### **Información proyecto - Carga de resultados de una `API` a `CSV` con Pandas y Python**

En este proyecto vamos a cargar los resultados de una `API` y los vamos a exportar a un archivo `CSV`. Se realizaran los siguientes pasos: 

- Obtuvimos una respuesta de la `API` y la convertimos en una respuesta `JSON`
- Creamos un `Dataframe` en base a la `respuesta JSON`
- Exportamos el `Dataframe` hacia un archivo `CSV` omitiendo los indices
