### **Obtener información de la API generada por el sitio `od-api.oxforddictionaries.com:443/api/v2/entries` utilizando `API token`**

Utilizamos la siguiente API:

https://od-api.oxforddictionaries.com:443/api/v2/entries/

#### **Conexión a la API**

```python
import requests
import json

app_id = "3410023e"
app_key = "6f82faaee87b2b823916b58f72f648b6"
language = "en-gb"
word_id = "hello"

url = "https://od-api.oxforddictionaries.com:443/api/v2/entries/" + language + "/" + word_id.lower()

# Agregamos el token en "headers"
response = requests.get(url, headers={"app_id": app_id, "app_key": app_key})

print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

#### **Extraer información en el tipo `diccionario`**

Se busca el significado de una palabra.

```python
r = json.loads(response.text)
print(r)
```
