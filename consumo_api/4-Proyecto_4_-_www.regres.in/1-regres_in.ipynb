### **`Obtener`, `enviar`, `actualizar` y `eliminar` datos de la API generada por el sitio `reqres.in/api/users`**

Utilizamos la siguiente API:

https://reqres.in/api/users

#### **Conexión a la API**

```python
import requests
import json
from pprint import pprint

url = "https://reqres.in/api/users"

def req(tipo, url, data={}):
    response = requests.request(tipo, url=url, data=data)
    return json.loads(response.text)
```

#### **`Paso 1: obtener información (GET)`**

Obtener toda la información de los usuarios retornada por la API y guardarla en una variable llamada **`users_data`** e imprimirla en pantalla

```python
users_data = req('GET', url)
pprint(users_data)
```

```python
objeto = users_data['data']
pprint(objeto)
```

Obtenemos los `id` de los usuarios:

```python
[elemento['id'] for elemento in objeto]
```

```python
lista = []
for elemento in objeto:
    lista.append(elemento['id'])

lista 
```

```python
[elemento['id'] for elemento in objeto if elemento['first_name'] == 'Tracey']
```

```python
lista = []
for elemento in objeto:
    if elemento['first_name'] == 'Tracey':
        lista.append(elemento['id'])

lista 
```

#### **`Paso 2: agregar (enviar) datos (POST)`**

Crearemos un usuario que tenga de nombre **`Ignacio`** y de trabajo **`Profesor`**. Guardaremos el diccionario de respuesta en una variable llamada **`created_user`** y luego imprimirla en pantalla

```python
post_data = {"first_name":"Ignacio",
             "job":"profesor"}

created_user = req('POST', url, data=post_data)
pprint(created_user)
```

#### **`Paso 3: actualizar datos (PUT)`**

Actualizaremos el usuario llamado **`George`** de **` id  =  1`** para que tenga un campo llamado **`residence`** igual a **`zion`**. Guardaremos el diccionario de respuesta en una variable llamada **`updated_user`** y luego la imprimiremos en pantalla

```python
put_data = {"email": "george.bluth@reqres.in",
            "first_name": "George",
            "last_name": "Bluth",
            "avatar": "https://reqres.in/img/faces/1-image.jpg",
            "residence":"zion"}

update_user = req('PUT', url+'/1', data=put_data)
pprint(update_user)
```

#### **`Paso 4: eliminar datos (DELETE)`**

Eliminaremos al usuario llamado **`Tracey`**. Imprimiremos el código de respuesta en pantalla

```python
import requests
import json
from pprint import pprint

url = "https://reqres.in/api/users"

def del_req(url, id):
    url = url + f'/{id}'
    print(url)
    return requests.request('DELETE', url=url)
```

De esta manera borramos el **`id = 6`** de **`Tracey`**

```python
print(del_req(url, 6))
```
