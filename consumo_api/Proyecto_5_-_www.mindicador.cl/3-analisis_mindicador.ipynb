### **`Prueba 1`**

#### **Conexión a la API**

```python
import requests
import json
import pandas as pd
from pprint import pprint

response = requests.get("https://mindicador.cl/api/")
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

#### **Extraer información en el tipo `string`**

```python
print(response.text)
print(type(response.text)) # <class 'str'>
```

#### **Extraer información en el tipo `diccionario`**

```python
pprint(response.json())
print(type(response.json())) # <class 'dict'>
```

```python
respuesta = json.loads(response.text)
pprint(respuesta)
print(type(respuesta)) # <class 'dict'>
```

##### **Prueba 2 orden y modificación columna `fecha`**

##### **Prueba 3 orden y modificación columna `fecha`**

Con esta opción solo damos formato a la fecha sin cambiar el tipo de dato, con el cual mantenemos el tipo de dato `object`, el cual no nos permite ordenar el DF según la columna `fecha`:

```python
import json
import requests
import pandas as pd
import datetime as dt

url = f'https://mindicador.cl/api/uf/2024'
response = requests.get(url)
data = json.loads(response.text.encode("utf-8"))
df = pd.json_normalize(data, record_path='serie')
df['fecha']= pd.to_datetime(df['fecha']).dt.strftime('%d-%m-%Y')
df
```
