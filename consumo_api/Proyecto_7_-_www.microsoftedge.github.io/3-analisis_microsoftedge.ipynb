### **Obtener información de la API generada por el sitio `microsoftedge.github.io/Demos`**

#### **Conexión a la API**

```python
import requests
import json
import pandas as pd

response = requests.get("https://microsoftedge.github.io/Demos/json-dummy-data/64KB.json")
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

Vamos a utilizar la libreria **`urllib`** para conectarnos a la API:

```python
import json
import urllib.request
import pandas as pd
from pprint import pprint

response = urllib.request.urlopen('https://microsoftedge.github.io/Demos/json-dummy-data/64KB.json')
```

#### **Extraer información en el tipo `lista`**

```python
cuerpo_respuesta = response.read()

json_respuesta = json.loads(cuerpo_respuesta.decode('utf-8'))
pprint(json_respuesta)
print(type(json_respuesta)) # <class 'list'>
```

#### **Normalizar data en un Dataframe**

```python
df = pd.json_normalize(json_respuesta)
df
```

#### **Exportar Dataframe en un archivo `CSV`**

```python
df.to_csv("microsoftedge.csv", index=False)
```
