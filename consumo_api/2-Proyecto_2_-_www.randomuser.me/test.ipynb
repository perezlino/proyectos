```python
import json
import requests
import pandas as pd
import json

class Github_Api():

    def __init__(self):
        pass

    def devolverDataframe(self):
        url = 'https://raw.githubusercontent.com/zyrikby/website_related/master/2021-04-flattening-json-using-pandas/vertx_example.json'
        response = requests.get(url)
        data = json.loads(response.text.encode("utf-8"))                
        df = pd.json_normalize(data, record_path=['matches', 'segments'], record_prefix='segments.',
                                     meta=['source_uid', 'media_type', 'source_path', 'status', 
                                            ['matches', 'uid'], 
                                            ['matches', 'artist'],
                                            ['matches', 'album'],
                                            ['matches', 'title'],
                                            ['matches', 'type'],
                                            ['matches', 'imdb_id'],
                                            ['matches', 'year']],
                                     errors='ignore')
        return df
    
    def exportarCSV(self):
        self.devolverDataframe().to_csv("vertx_data2.csv", index=False)
```

```python
api1 = Github_Api()
api1.devolverDataframe()
```

```python
api1.exportarCSV()
```
