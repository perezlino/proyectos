### **Obtener información de la API generada por el sitio `randomuser.me/api` y normalizarla en un Dataframe**

Utilizamos la siguiente API:

https://randomuser.me/api/

#### **Conexión a la API**

```python
import requests
import json
from pandas import json_normalize

response = requests.get("https://randomuser.me/api/")
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

#### **Extraer información en el tipo `string`**

```python
print(response.text)
print(type(response.text)) # <class 'str'>
```

#### **Extraer información en el tipo `diccionario`**

```python
print(response.json())
print(type(response.json())) # <class 'dict'>
```

```python
respuesta = json.loads(response.text)
print(respuesta)
print(type(respuesta)) # <class 'dict'>
```

#### **Extraer información de la API**

Utilizamos el diccionario **`respuesta`** creado anteriormente:

```python
from pprint import pprint

pprint(respuesta['results'])
print(type(respuesta['results'])) # <class 'list'>
```

```python
pprint(respuesta['results'][0])
print(type(respuesta['results'][0])) # <class 'dict'>
```

#### **Normalizar la data en un Dataframe**

##### **`Forma 1`**

```python
answer = respuesta['results'][0]

df = json_normalize({
    'firstname':answer['name']['first'],
    'lastname':answer['name']['last'],
    'country':answer['location']['country'],
    'username':answer['login']['username'],
    'password':answer['login']['password'],
    'email':answer['email']
})

print(df)
print(type(df)) # <class 'pandas.core.frame.DataFrame'>
```

```python
display(df)
```

##### **`Forma 2`**

```python
import pandas as pd

flattened_data = []
for e in respuesta['results']:
    nat = e.get('nat', pd.NA)
    phone = e.get('phone', pd.NA)
    gender = e.get('gender', pd.NA)
    email = e.get('email', pd.NA)
    cell = e.get('cell', pd.NA) 
    # age = e['dob]['age] <------------- Lo mismo
    age  = e.get('dob', pd.NA)['age']      
    date  = e.get('dob', pd.NA)['date']  
    name  = e.get('id', pd.NA)['name']      
    value  = e.get('id', pd.NA)['value']             
    entry = {
                        'nat': nat,
                        'telefono': phone,
                        'genero': gender,
                        'correo': email,
                        'cell': cell,
                        'edad': age,
                        'fecha': date,
                        'nombre': name,
                        'valor': value,
                        }
    flattened_data.append(entry)
    
    print(flattened_data)
    
df = pd.DataFrame(flattened_data)
```
