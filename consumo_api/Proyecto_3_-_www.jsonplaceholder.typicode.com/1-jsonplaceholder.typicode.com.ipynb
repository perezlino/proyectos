### **Obtener información de la API generada por el sitio `jsonplaceholder.typicode.com/users`, normalizarla en un Dataframe y exportarla a un archivo CSV**

Utilizamos la siguiente API:

https://jsonplaceholder.typicode.com/users

#### **Conexión a la API**

```python
import requests
import json
import pandas as pd

response = requests.get("https://jsonplaceholder.typicode.com/users")
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

#### **Extraer información en el tipo `string`**

```python
print(response.text)
print(type(response.text)) # <class 'str'>
```

#### **Extraer información en el tipo `lista`**

```python
print(response.json())
print(type(response.json())) # <class 'dict'>
```

```python
respuesta = json.loads(response.text)
print(respuesta)
print(type(respuesta)) # <class 'dict'>
```

#### **Normalizar data en un Dataframe**

##### **`Forma 1`**

Utilizamos el diccionario **`respuesta`** creado anteriormente:

```python
flattened_data = []
for a in respuesta:
    id = a['id']
    name = a['name']
    username = a['username']
    email = a['email']
    street = a['address']['street']
    suite = a['address']['suite']
    city = a['address']['city']
    zipcode = a['address']['zipcode']
    lat = a['address']['geo']['lat']
    lng = a['address']['geo']['lng']
    phone = a['phone']
    website = a['website']
    entry = {
                'id': id,
                'nombre': name,
                'nombre_de_usuario': username,
                'correo': email,
                'direccion': street,  
                'suite': suite,
                'ciudad': city,
                'zipcode': zipcode,
                 'latitud': lat,
                 'longitud': lng,
                 'telefono': phone,
                 'website': website,
            }
    flattened_data.append(entry)
    
print(flattened_data)
```

```python
df = pd.DataFrame(flattened_data)
df
```

##### **`Forma 2`**

```python
df = pd.DataFrame(respuesta)
df
```

```python
new_df = pd.json_normalize(respuesta)
new_df
```

#### **Exportar Dataframe en un archivo `CSV`**

```python
df.to_csv("users.csv", index=False)
```
