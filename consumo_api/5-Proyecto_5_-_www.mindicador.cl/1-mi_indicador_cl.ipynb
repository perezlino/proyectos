### **`Prueba 2`**

```python
import json
import requests
import pandas as pd
import json
import datetime as dt

class Midicador():

    def __init__(self, opcion, indicador, fecha):
         self.opcion = opcion
         self.indicador = indicador
         self.fecha = fecha

    def InfoApi(self):
            if self.opcion == 1:
                url = f'https://mindicador.cl/api/{self.indicador}'
                response = requests.get(url)
                data = json.loads(response.text.encode("utf-8"))                
                df = pd.json_normalize(data, record_path='serie', meta=['codigo','nombre','unidad_medida'])
                df['fecha']= pd.to_datetime(df['fecha']).dt.strftime('%d-%m-%Y')
                return df
            elif self.opcion == 2:    
                # Podemos indicador solo el año o la fecha (dia-mes-año)
                url = f'https://mindicador.cl/api/{self.indicador}/{self.fecha}'
            response = requests.get(url)
            data = json.loads(response.text.encode("utf-8"))
            df = pd.json_normalize(data, record_path='serie')
            df['fecha']= pd.to_datetime(df['fecha']).dt.strftime('%d-%m-%Y')
            return df
```

#### **Obtener los valores de un indicador especifico para los ultimos 30 dias**

```python
indicadores1 = Midicador(1,'uf','')
indicadores1.InfoApi()
```

#### **Obtener los valores de un indicador para un año especifico**

```python
indicadores2 = Midicador(2,'dolar','2023')
indicadores2.InfoApi()
```
