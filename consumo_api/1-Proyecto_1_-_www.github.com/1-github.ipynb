### **Obtener informacion de archivo JSON desde un repositorio de Github, normalizarla en un Dataframe y exportarla a un archivo CSV**

### **`Prueba 1`**

#### **Conexión a la API**

Utilizamos el siguiente enlace:

https://raw.githubusercontent.com/zyrikby/website_related/master/2021-04-flattening-json-using-pandas/vertx_example.json

```python
import requests
import json
import pandas as pd

response = requests.get("https://raw.githubusercontent.com/zyrikby/website_related/master/2021-04-flattening-json-using-pandas/vertx_example.json")
print(response.status_code) # 200 ==> Quiere decir que la conexión es satisfactoria
```

#### **Extraer información en el tipo `lista`**

```python
print(response.json())
print(type(response.json())) # <class 'list'>
```

```python
respuesta = json.loads(response.text)
print(respuesta)
print(type(respuesta)) # <class 'list'>
```

#### **Extraer información de la API y normarlizar la data en un Dataframe**

##### **`Forma 1`**

```python
def load_vertx_data():
    flattened_data = []
    for mte in respuesta:
        que_uid = mte.get('source_uid', pd.NA)
        media_type = mte.get('media_type', pd.NA)
        source_path = mte.get('source_path', pd.NA)
        status = mte.get('status', pd.NA)
        for m in mte['matches']:
            artist = m.get('artist', pd.NA)
            album = m.get('album', pd.NA)
            title = m.get('title', pd.NA)
            type_ = m.get('type', pd.NA)
            ref_uid = m.get('uid', pd.NA)
            imdb_id = m.get('imdb_id', pd.NA)
            year = m.get('year', pd.NA)
            for s in m['segments']:
                duration = s['duration']
                que_offset = s['que_offset']
                ref_offset = s['ref_offset']
                entry = {
                        'que_uid': que_uid,
                        'media_type': media_type,
                        'source_path': source_path,
                        'status': status,
                        'artist': artist,
                        'album': album,
                        'title': title,
                        'type': type_,
                        'ref_uid': ref_uid,
                        'imdb_id': imdb_id,
                        'year': year,
                        'que_offset': que_offset,
                        'ref_offset': ref_offset,
                        'duration': duration,
                        }
                flattened_data.append(entry)
    
    print(flattened_data)
    
    return pd.DataFrame(flattened_data)
```

```python
df = load_vertx_data()
df
```

##### **`Forma 2`**

```python
def cargar_vertx_data_usando_json_normalize():
    df = pd.json_normalize(
        respuesta,
        record_path=['matches', 'segments'],
        meta=['source_uid', 'media_type', 'source_path', 'status', 
              ['matches', 'uid'], 
              ['matches', 'artist'],
              ['matches', 'album'],
              ['matches', 'title'],
              ['matches', 'type'],
              ['matches', 'imdb_id'],
              ['matches', 'year'],
        ],
        errors='ignore'
    )
    df = df.rename(
        columns={
            'source_uid': 'que_uid', 
            'matches.uid': 'ref_uid',
            'matches.artist': 'artist',
            'matches.album': 'album',
            'matches.title': 'title',
            'matches.type': 'type',
            'matches.imdb_id': 'imdb_id',
            'matches.year': 'year',
        },
    )
    
    return df
```

```python
new_df =  cargar_vertx_data_usando_json_normalize()
new_df
```

#### **Exportar Dataframe en un archivo `CSV`**

```python
new_df.to_csv("vertx_data.csv", index=False)
```

### **`Prueba 2`**

#### **Crear objetos y funciones que faciliten la llamada a la API**

```python
import json
import requests
import pandas as pd
import json

class Github_Api():

    def __init__(self):
        pass

    def devolverDataframe(self):
        url = 'https://raw.githubusercontent.com/zyrikby/website_related/master/2021-04-flattening-json-using-pandas/vertx_example.json'
        response = requests.get(url)
        data = json.loads(response.text.encode("utf-8"))                
        df = pd.json_normalize(data, record_path=['matches', 'segments'], record_prefix='segments.',
                                     meta=['source_uid', 'media_type', 'source_path', 'status', 
                                            ['matches', 'uid'], 
                                            ['matches', 'artist'],
                                            ['matches', 'album'],
                                            ['matches', 'title'],
                                            ['matches', 'type'],
                                            ['matches', 'imdb_id'],
                                            ['matches', 'year']],
                                     errors='ignore')
        return df
    
    def exportarCSV(self):
        self.devolverDataframe().to_csv("vertx_data2.csv", index=False)
```

```python
api1 = Github_Api()
```

```python
api1.devolverDataframe()
```

```python
api1.exportarCSV()
```
