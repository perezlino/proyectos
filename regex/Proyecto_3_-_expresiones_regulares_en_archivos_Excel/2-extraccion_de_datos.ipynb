### **Extracción de datos desde el libro de Excel**

#### **`Enunciado 1`: Queremos obtener los apellidos de todos los empleados cuyos salarios son inferiores a treinta mil, pero al menos 24000 dólares. Así que esto significa 24000 hasta 29999**

##### **Lectura del libro de Excel**

```python
import re
from pprint import pprint
from openpyxl import load_workbook

try:
    workbook = load_workbook(filename="empleados.xlsx")
    sheet = workbook["EmployeeData"]
except Exception as e:
    print(e)  
```

##### **Preparar los datos**

Añadir los datos como string a una lista:

```python
#Crear la lista vacía para almacenar los datos
data = []

#Iterando sobre sheet.values y añadiendo los datos como strings
for row in sheet.values:
    a, b, c, d, e, f, g = row
    data.append("{};{};{};{};{};{};{}".format(a, b, c, d, e, f, g))

pprint(data)
```

```python
#Joining los strings en la lista data usando el \n como delimitador.
empleados = "\n".join(data)

print(empleados)
```

##### **Test**

```python
# El primer ".+" no respeta el orden de los ; ya que, no cancelamos la codicia del + y avanza lo más que puede hasta que encontró el grupo 
# que corresponderia al antepenúltimo valor, dado que, despúes de él aún hay otros dos valores
s1 = re.findall(r"\d{1,};.+;(.+?);.+;2[4-9]\d{3}", empleados)
s1
```

##### **Desarrollo**

Ahora, después de este grupo, tenemos otro punto y coma y luego usé el punto y el signo más (.+), pero, esta vez, como puedes notar aquí, no he usado el signo de interrogación. ¿Por qué?, eso es porque esta vez quiero que el signo más (+) sea tan codicioso como sea posible y coincida con todo el texto hasta el último punto y coma en cada línea.

```python
# Con el primer ".+?" respetamos el orden de los ; ya que cancelo la codicia del + 
s1 = re.findall(r"\d{1,};.+?;(.+?);.+;2[4-9]\d{3}", empleados)
s1
```
