### **4.3 - Proyecto 3 - `Logs`**

#### **Pregunta 1:** `Queremos todas las entradas de log que sean "críticas" y se hayan generado entre los días "11 y 16 del mes de Enero de 2020" y extraer la fuente de cada entrada, es decir, el componente que causó el problema`

En primer lugar tenemos el "nivel de gravedad del mensaje", normalmente puede ser un error crítico, advertencia, informativo, etc. Luego tenemos un "espacio". A continuación tenemos la "fecha y hora" en que se generó el mensaje de log. Separado por un espacio después de la hora tenemos la "fuente", que es el componente de software que generó el mensaje de log. Por lo tanto, en nuestro caso sería la fuente, seguida de otra información como el "ID del evento" y una breve "descripción del problema". 

Ahora, con este mapa de cada entrada de log en mente podemos empezar a escribir patrones de expresiones regulares de acuerdo a nuestras necesidades. Supongamos que `queremos buscar todas las entradas de log que sean "críticas" y se hayan generado entre los días "11 y 16 del mes de Enero de 2020" y extraer la parte de la fuente de cada entrada, es decir, el componente que causó el problema`.

```python
import re

try:
    archivo = open('logs.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

```python
#           Nivel de gravedad      Fecha    Hora [A-P/M] Fuente ...
result = re.findall(r"Critical 1/1[1-6]/2020 .+ [A-Z]{2} (.+?) \d+", string)
result
```

#### **Pregunta 2:** `Queremos todas las entradas de log que se hayan generado entre las 12, 1, 2 o 3 PM y extraer la fuente de cada entrada, es decir, el componente que causó el problema`

```python
import re

try:
    archivo = open('logs.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Testeo**

```python
#      Nivel de gravedad Fecha    Hora            [A-P]M Fuente ...
result = re.findall(r".+/2020 12|[1-3]:\d\d:\d\d PM (.+?) \d+", string)
result
```

##### **Desarrollo**

```python
# Los caracteres ?: generan un "Grupo no capturable"
#      Nivel de gravedad Fecha    Hora            [A-P]M Fuente ...
result = re.findall(r".+/2020 (?:12|[1-3]):\d\d:\d\d PM (.+?) \d+", string)
result
```

#### **Pregunta 3:** `Queremos todas las entradas de log que hayan sido generadas por la fuente TPM y extraer la fecha`

##### **Desarrollo**

#### **Pregunta 4:** `Queremos todas las entradas de log que se hayan generado entre los días "24 y 27 del mes de Enero de 2020" entre las 8 AM y 9 AM y extraer fecha y hora`
