#### **Pregunta 1:** `Queremos hacer coincidir todas las URL del archivo que corresponden a sitios web de compras en línea`

En primer lugar, supongamos que queremos escribir un patrón que coincida con todas las URL del archivo asociadas a servicios de compra en línea. Para ello, primero tenemos que analizar la estructura de una URL. Así, mirando la primera entrada, la primera URL del archivo, tenemos "https". Es el protocolo web seguido de dos puntos y dos barras inclinadas. Además, fíjate que en algunos casos tenemos un subdominio listado como este de aquí, "login.tmall.com". Además, tenga en cuenta que aunque en nuestra lista sólo tenemos sitios web "https", algunos sitios web antiguos pueden utilizar el protocolo no secular heredado que es el "http".

```python
import re

try:
    archivo = open('web.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

Ahora, como siempre, echemos un vistazo a la estructura general de cada línea de cada entrada de este archivo. En primer lugar, cada línea comienza con el nombre real del sitio web. Por ejemplo, Google, YouTube, Facebook o Alipay. Así pues, se trata de una colección de caracteres que podemos representar fácilmente utilizando los metacaracteres como el punto y el signo más (.+) para señalar la aparición de al menos un carácter, cualquier carácter excepto la nueva línea. Luego, el nombre de cada sitio web va seguido de uno o más espacios en blanco. Ahora bien, estos espacios en blanco pueden ser en realidad uno o varios espacios normales o uno o más tabuladores (\t), por ejemplo. En cualquier caso, podemos señalar su presencia utilizando la secuencia especial \s que puede repetirse una o más veces, de ahí el signo más que sigue a \s en mi patrón. A continuación, tenemos la propia URL que tenemos que hacer coincidir de acuerdo con nuestra tarea. Como ya hemos comentado al analizar la estructura de una nueva URL, el protocolo puede ser "http" o "https". Por lo tanto, utilizamos "http" y luego añadimos la opción de tener otro carácter antes de los dos puntos. Esto coincidiría con la letra "s" cuando se tiene "https" como protocolo. Sin embargo, note que usé el asterisco que significa 0 o más ocurrencias del carácter precedente, esto que justo aquí después de "http". En otras palabras, esto significa que podemos o no tener este carácter adicional después de "http" y así es como podemos cubrir ambos casos con la misma sintaxis simple. A continuación, tenemos los dos puntos seguidos de dos barras inclinadas y luego tenemos el nombre de dominio. Por ejemplo, podríamos tener "Alipay" o "Twitch" o "Google" o "Facebook" y así sucesivamente. Podemos hacer coincidir este nombre, el nombre del dominio, utilizando el punto y el signo más (.+), ya que se trata de una colección de al menos un carácter, cualquier carácter, excepto la nueva línea. Y eso es exactamente lo que hice aquí. Siguiendo adelante, después del nombre de dominio tenemos el punto, justo aquí, un punto real para la extensión del dominio como ".com", ".us", ".tv" y así sucesivamente. Esto se puede hacer coincidir utilizando dos o más caracteres alfanuméricos consecutivos. Fíjese en la secuencia especial \w y en el dos seguido de una coma entre llaves, que significa dos o más apariciones de la secuencia \w. Necesitamos dos o más porque hay extensiones de dos letras como ".us" o ".tv", así como extensiones de tres letras como ".com" o ".org". Ahora, para que coincida con toda la URL necesitamos incluirla entre paréntesis para crear un grupo. Así pues, este es el grupo que señala esa URL, al que sigue otro carácter de espacio en blanco seguido de muchos otros caracteres hasta el final de la línea. Entre estos caracteres nos interesa que las palabras "online shopping" formen parte de esta cadena, ya que nuestra tarea consiste en identificar las URL de los sitios web que ofrecen servicios de compra en línea.

```python
result = re.findall(r".+\s+(http.*://.+\.\w{2,})\s.+Online shopping.+", string)
result
```

#### **Pregunta 2:** `Queremos hacer coincidir el nombre del sitio cuyos websites terminen en ".com" o ".org" y sean de China `

```python
import re

try:
    archivo = open('web.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

```python
result = re.findall(r"(.+)\s+http.*://.+\.\w{3}\s.+China.+", string)
result
```

#### **Pregunta 3:** `Queremos hacer coincidir todas las URL y su pais de origen donde su dominio contenga entre 1 y 4 caracteres y corresponda a una red social`

```python
import re

try:
    archivo = open('web.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

```python
result = re.findall(r".+\s+(http.*://.{1,4}\.\w{2,})\s.+Social networking\s(.+?)\s.+", string)
result
```

#### **Pregunta 4:** `Queremos hacer coincidir el nombre del sitio y su protocolo para todas aquellas URL que corresponden a un subdominio`

```python
import re

try:
    archivo = open('web.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```
