### **4.2 - Proyecto 2 - `Guía telefónica`**

#### **Pregunta 1:** `Queremos hacer coincidir el apellido y el número de teléfono de todas las personas cuyo código de área termine en cero`

Estudiemos la estructura de cada entrada de la guía telefónica. Así, en primer lugar observamos que cada entrada está en una línea separada. A continuación, cada línea empieza con el "nombre" y el "apellido" de la persona, seguido de un carácter de tabulación (\t). el caracter tabulador es seguido por el código de área entre paréntesis, un espacio y luego el número de teléfono. Así que ese sería el mapa general por así decirlo de cada entrada en la guía telefónica.

Supongamos que de esta colección de nombres y números de teléfono queremos obtener `el apellido y el número de teléfono de todas las personas cuyo código de área termine en cero`. Así que vamos a necesitar tener dos grupos que deben coincidir en el string objetivo. El primer grupo coincidirá con el apellido y el segundo grupo coincidirá con el número de teléfono excluyendo el código de área. Esto significa que la única condición que debe cumplir cada entrada válida es tener un código de área que termine en cero.

```python
import re

try:
    archivo = open('guia_telefonica.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Testeo**

```python
# \( --> escapa el parentesis y \) --> escapa el paréntesis de cierre. Esto es para no usar los () como delimitadores de grupo
# y utilizarlos como parte del string.
string = "Branden Cantu	(862) 408-5783\nPatricia Dodson	(685) 849-8192\nAyaan Novak	(812) 883-7304"
result = re.findall(r"\(\d{2}[2]\)", string)
result
```

```python
string = "Branden Cantu	(862) 408-5783\nPatricia Dodson	(685) 849-8192\nAyaan Novak	(812) 883-7304"
result = re.findall(r"\(\d{2}2\)", string)
result
```

##### **Desarrollo**

```python
result = re.findall(r".+? (.+?)\t\(\d{2}[0]\)\s(.*)", string)
result
```

```python
result = re.findall(r".+ (.+)\t\(\d\d0\)\s(.+)", string)
result
```

#### **Pregunta 2:** `Queremos hacer coincidir el área de cada número de teléfono que termine en siete`

Supongamos que `queremos obtener el código de área de cada número de teléfono que termine en siete`. Bien, en este caso necesitamos definir un único grupo dentro del patrón, el que representa el código de área.

```python
import re

try:
    archivo = open('guia_telefonica.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

```python
# Es erróneo
result = re.findall(r".+\t\((\d{3})\)\s.+7", string)
result
```

```python
# Es erróneo
result = re.findall(r".+\t\((\d{3})\) .+7", string)
result
```

```python
# Es erróneo
result = re.findall(r".+\t\((\d{3})\) .+[7]", string)
result
```

```python
result = re.findall(r".+\t\((\d{3})\)\s.{7}7", string)
result
```

#### **Pregunta 3:** `Queremos hacer coincidir el nombre y apellido de cada cliente donde su número de teléfono comience con 1, 3, 5, 7 o 9`

##### **Desarrollo**

```python
result = re.findall(r"(.+)\t\(\d{3}\) [13579].{7}", string)
result
```

```python
result = re.findall(r"(.+)\t\(\d{3}\) [13579].+", string)
result
```

#### **Pregunta 4:** `Queremos hacer coincidir el nombre de cada cliente donde el área del teléfono comience con 0, 1 o 2`

##### **Desarrollo**

```python
result = re.findall(r"(.+) .+\t\([0-2]\d\d\) .+", string)
result
```

```python
result = re.findall(r"(.+) .+\t\([0-2]\d\d\) .{8}", string)
result
```

```python
result = re.findall(r"(.+) .+[aeiou]\t\(\d{3}\) .{7}[079]", string)
result
```
