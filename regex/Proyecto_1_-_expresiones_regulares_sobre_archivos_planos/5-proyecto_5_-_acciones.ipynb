### **4.5 - Proyecto 5 - `Acciones`**

#### **Pregunta 1:** `Queremos hacer coincidir todos los nombres de empresas de esta lista donde su ingreso sea inferior a 50 billones de USD`

Este archivo contiene diversa información sobre 30 acciones estadounidenses, como el nombre de la empresa, el volumen medio negociado, los ingresos de cada empresa y también el ratio P/E, también llamado ratio precio/beneficios. `Ahora, se nos pide que hagamos coincidir todos los nombres de empresas de esta lista y sus ingresos, siempre que éstos sean inferiores a 50.000 millones de USD (50 billones de USD)`.

Ahora, fijémonos en un par de cosas sobre cada entrada de esta lista. Después del nombre de la empresa en cada línea tenemos tres valores correspondientes al "Volumen medio" expresado en millones. Luego tenemos los "Ingresos", que se expresan en billones, fijémonos en la B del final. Y por último tenemos el "Ratio P/E", que de nuevo es un número. Así que tenemos tres números flotantes en cada línea.

```python
import re

try:
    archivo = open('acciones.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

##### **Desarrollo**

En primer lugar, está el nombre de la empresa que queremos que coincida, ya que cada línea comienza con un nombre de empresa y queremos que coincida con eso. Definí el grupo que contiene uno o más caracteres, cualquier carácter excepto la nueva línea que aparece al menos una vez para señalar el nombre de la empresa. Justo después del nombre de la empresa en cada línea, tenemos uno o más caracteres consecutivos de espacio en blanco que cubrimos utilizando la secuencia especial \s y el metacaracter de signo más (+). A continuación tenemos el primer número correspondiente al "Volumen medio de negociación". Se trata de un número de coma flotante, es decir, un número con decimales, y lo emparejamos utilizando la clase de caracteres 0 a 9 que aparecen una o más veces. A continuación, el punto escapado (\.) seguido de otra serie de uno o más dígitos y la letra M mayúscula de millones. Si miramos dentro de nuestro archivo, tenemos otra secuencia de uno o más espacios en blanco seguidos del segundo número de la línea, que es el "Ingreso". Este número tiene que cumplir nuestra condición, es decir, tiene que ser inferior a 50. Esta es la razón por la que el primer dígito del patrón puede ser cualquier cifra del 1 al 4, como puede ver aquí con esta clase de caracteres, mientras que el segundo dígito puede ser cualquier cifra del 0 al 9. Esto permite números de hasta 49, lo que sigue cumpliendo nuestra condición. Luego tenemos el punto escapado (\.). Y luego uno o más decimales seguidos de B de billones. Luego tenemos otra secuencia de uno o más espacios en blanco, justo aquí \s más (\s+) justo antes del tercer y último número en cada línea que es otro punto flotante a número correspondiente al "Valor de la relación P/E".

```python
result = re.findall(r"(.+?)\s+[0-9]+\.[0-9]+M\s+[1-4][0-9]\.[0-9]+B\s+[0-9]+\.[0-9]+", string)
result
```
