#### **Pregunta 1:** `Queremos hacer coincidir todos los autores cuyos títulos de libros tengan menos de 25 caracteres.`

Fijémonos en un par de cosas que nos ayudarán a realizar casi todas nuestras tareas de coincidencia de patrones. ¿Qué sabemos que es cierto con respecto a todo este string objetivo que definimos? En primer lugar, cada entrada, es decir, el nombre del autor, el título del libro y el año, se encuentra en su propia línea, lo que significa que las entradas están separadas por un carácter de nueva línea **`\n`**. Esto actuará como el borde exterior de cada entrada en nuestra base de datos de la librería. En segundo lugar, sabemos que en cada línea el nombre del autor, el título del libro y el año están separados por un borde interior que es el punto y coma (**`;`**). Estos bordes o límites nos ayudarán enormemente a realizar nuestras tareas de concordancia de patrones.

```python
import re

try:
    archivo = open('libros.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

En primer lugar, sabemos que tenemos un grupo de caracteres que representan el "nombre del autor" justo antes del primer punto y coma de cada línea. No nos interesa que coincidan esos caracteres, pero conocemos su ubicación en cada entrada. Tras el primer punto y coma tenemos el "título del libro" en cada línea. Aquí es donde debemos centrarnos y esto es lo que tenemos que hacer coincidir. Por lo tanto, he utilizado el metacaracter de punto (.) para hacer referencia a un caracter, cualquier caracter excepto la nueva linea, y luego utilizando llaves he especificado el limite inferior y superior de caracteres que debe tener nuestra coincidencia y que esta entre 1 y 25 caracteres inclusive. Por supuesto, como el "título del libro" es lo que estamos buscando, he incluido este sub-patrón en un grupo usando paréntesis. Inmediatamente después del título del libro, tenemos otro punto y coma seguido de otra cadena de caracteres que realmente no nos importa. Sería el "año" que aparece al final del patrón.

```python
result = re.findall(r".+?;(.{1,25});.+?", string)
result
```

##### **Desarrollo**

```python
result = re.findall(r".+;(.{1,25});.+", string)
result
```

#### **Pregunta 2:** `Queremos hacer coincidir todos los autores de esta base de datos que publicaron sus libros a partir del año 2000.`

```python
import re

try:
    archivo = open('libros.txt', 'r', encoding='UTF-8')
    string = archivo.read() 
except Exception as e:
    print(e)    
finally:
    archivo.close()

print(string)
```

En primer lugar, fijémonos en el grupo, que debe coincidir con el nombre del autor en cada línea que cumpla la condición relativa al año. Luego tenemos los dos puntos y coma con el título del libro en medio. Nada de interés aquí. Sólo observamos que aquí es donde está el título. Por último, he especificado la condición que debe cumplirse en cada línea, después del segundo punto y coma. Así, como queremos todos los autores que publicaron sus libros después del año 2000, los dos primeros caracteres que necesitamos son "2" y "0", obviamente, luego el tercer y cuarto dígito pueden ser cualquier cifra del 0 al 9. Por eso he utilizado dos clases de caracteres consecutivos que representan dígitos. Por supuesto, podría haber utilizado la secuencia especial \d en lugar de esta clase de caracteres y el resultado habría sido el mismo.

```python
result = re.findall(r"(.+);.+;20[0-9][0-9]", string)
result
```
