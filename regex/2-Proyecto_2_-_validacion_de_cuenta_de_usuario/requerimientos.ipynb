### **Requerimientos**

**`Primer nombre`**:
- No contiene otros caracteres además de letras.
- Empieza por mayúscula y todas las demás letras son minúsculas.
- Tiene al menos 2 caracteres.

**`Apellido`**:
- No contiene más caracteres que letras.
- Empieza por mayúscula y todas las demás letras son minúsculas.
- Tiene al menos 2 caracteres.

**`Fecha de nacimiento`**:
- La persona nació antes del 1 de enero de 2002.
- Sólo contiene dígitos y barras diagonales.
- Sigue el formato DD/MM/AAAA.

**`Dirección de correo electrónico`**:
- Sigue el formato correcto: username@domain.extension
- El nombre de usuario sólo puede contener letras, dígitos, puntos o el guión bajo.
- El dominio sólo puede contener letras minúsculas.
- La extensión sólo puede contener 2, 3 ó 4 letras minúsculas.

**`Nombre de usuario`**:
- No contiene otros caracteres además de letras, dígitos o el guión bajo.
- Tiene un mínimo de 6 caracteres y un máximo de 12 caracteres.

**`Contraseña`**:
- Tiene al menos 8 caracteres.
- Empieza por una letra minúscula.
- Debe contener al menos una letra minúscula, una mayúscula, un dígito y uno de los siguientes caracteres: $, &, ?, ! o %.

**`Número de tarjeta de crédito`**:
- La tarjeta debe ser emitida únicamente por Visa o Mastercard (empezando por un 4 o un 5).
- Tiene exactamente 16 dígitos y no contiene otros caracteres aparte de los dígitos.

**`Fecha de caducidad de la tarjeta de crédito`**:
- No es anterior a mayo de 2024 inclusive.
- Sólo contiene dígitos y barras diagonales.
- Sigue el formato MM/AA

**`Código de verificación de la tarjeta de crédito / valor de verificación (CVC / CVV)`**:
- Es un número de 3 dígitos, no contiene ningún otro carácter.


**~ Sugerencias para resolver algunas de las tareas del proyecto ~**

- **`Fecha de nacimiento`**: En cuanto a la fecha de nacimiento - la aplicación valida este valor sólo si la fecha es anterior al 1 de enero de 2002, en el formato requerido, es decir, 31/12/2001 por ejemplo. Así, cualquier mes del 01 al 12 es válido; cualquier día del 01 al 31 es válido; y cualquier año anterior al 2002 es válido. Por lo tanto, para especificar cualquiera de estos rangos, necesitará utilizar clases de caracteres y operaciones OR, utilizando el metacaracter "pipe". No olvide las barras diagonales que separan el mes, el día y el año.

- **`La contraseña`**: La contraseña debe empezar por una letra minúscula (clase de caracteres aquí) y tener al menos 8 caracteres, sin límite superior. Para especificar los tipos de caracteres que debe contener la contraseña, yo utilizaría clases de caracteres. Además, para indicar que la contraseña debe contener cada una de estas clases (letras minúsculas, letras mayúsculas, dígitos y símbolos especiales), utilizaría varias afirmaciones positivas para cada clase de caracteres.

- **`Fecha de caducidad de la tarjeta de crédito`**: Después de escribir el patrón para la fecha de nacimiento, el de validación de la fecha de caducidad de la tarjeta de crédito debería ser un poco más fácil de concebir. No obstante, se trata de un patrón diferente que requiere especial atención. La tarjeta de crédito no debe caducar antes de mayo de 2024, y la fecha tiene que seguir el formato MM/AA - en nuestro caso sería 24/05. Por lo tanto, divida esta cuestión en dos - el resto del año 2024 a partir de mayo, y luego cualquier año posterior, sin límite superior.Por lo tanto, necesitará varias operaciones OR - tenga en cuenta sus metacaracteres "pipe", así como algún proceso de pensamiento sobre cómo hacer coincidir los meses 05-12 en 2024, y luego todos los demás meses a partir de 2025.
