### **Carga y manejo de un libro de Excel en Python**

```python
import re
from pprint import pprint
from openpyxl import load_workbook

try:
    workbook = load_workbook(filename="empleados.xlsx")
    sheet = workbook["EmployeeData"]
except Exception as e:
    print(e)    
```

Podemos visualizar las columnas que tiene la Hoja de calculo:

```python
sheet.dimensions
```

Queremos tener toda la información de esta hoja representada como una cadena para poder buscar patrones de expresiones regulares dentro de la cadena.

```python
for row in sheet.values:
    print(row)
```

```python
#Crear la lista vacía para almacenar los datos
data = []

#Iterando sobre sheet.values y añadiendo los datos como strings
for row in sheet.values:
    a, b, c, d, e, f, g = row
    data.append("{};{};{};{};{};{};{}".format(a, b, c, d, e, f, g))

pprint(data)
```

```python
#Joining los strings en la lista data usando el \n como delimitador.
employees = "\n".join(data)

print(employees)
```
