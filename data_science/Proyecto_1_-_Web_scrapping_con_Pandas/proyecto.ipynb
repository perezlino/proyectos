#### **Paso a paso: `Múltiples temporadas para 3 ligas de Inglaterra -> Premier League, Championship y Liga 1`**

Revisamos el sitio el cual scrapearemos y veemos que existen varios links de descarga de archivos CSV con los resultados de las distintas ligas de fútbol de Inglaterra:

<center><img src="https://i.postimg.cc/MHyJVFW9/dl50.png"></center>

#### **Desarrollo**

```python
import pandas as pd
import numpy as np
```

**Link: base + temporada + liga**

```python
# Estructura del link
"https://www.football-data.co.uk/mmz4281/" + "2324" + "/" + "E0" + ".csv"
```

```python
# crear una base variable
base = "https://www.football-data.co.uk/mmz4281/"
base
```

##### **Código**

```python
ligas = ['E0','E1','E2']

lista_df = []

# recorre el rango de temporadas e inserta una nueva columna
for liga in ligas:
    for temporada in range(19,24): # rango de temporadas
        df = pd.read_csv(base + str(temporada) + str(temporada+1) + "/" + liga + ".csv")
        df.insert(1, 'temporada', str(temporada)+"/"+str(temporada+1))
        lista_df.append(df)
```

```python
# dataframes totales (5 temporadas x 3 ligas)
len(lista_df)
```

```python
# Vamos a concantenar las temporadas 19/20, 20/21, 21/22, 22/23 y 23/24 de la Premier League
df_premier_league = pd.concat(lista_df[0:5])
df_premier_league
```

```python
# renombramos las siguientes columnas
df_premier_league = df_premier_league.rename(columns={'Date':'fecha',
                                                      'HomeTeam':'equipo_local',
                                                      'AwayTeam':'equipo_visita',
                                                      'FTHG': 'goles_local',
                                                      'FTAG': 'goles_visita',
                                                      'FTR': 'resultado'})
```

```python
# Queremos filtrar solo los partidos donde se vea involucrado Manchester City
df_man_city = df_premier_league.loc[df_premier_league.equipo_local=='Man City', ['temporada', 'fecha', 'equipo_local', 'equipo_visita', 'goles_local', 'goles_visita', 'resultado']]
df_man_city
```

```python
len(df_man_city)
```

```python
# generar una nueva columna con numeros escalares ordenados
indice_= np.arange(0,93)
indice_
```

```python
df_man_city['indice'] = indice_
df_man_city
```

```python
# establecer columna indice como índice y actualizamos el Dataframe
df_man_city.set_index('indice', inplace=True)
df_man_city
```

#### **Paso a paso: `Múltiples temporadas para La Liga (España), Bundesliga (Alemania) y la Premier League (Inglaterra)`**

#### **Desarrollo**

```python
import pandas as pd
```

**Link: base + temporada + liga**

```python
# Estructura del link
"https://www.football-data.co.uk/mmz4281/" + "2324" + "/" + "E0" + ".csv"
```

```python
# crear una base variable
base = "https://www.football-data.co.uk/mmz4281/"
base
```

```python
# crear un diccionario con el nombre original de la liga como key
dict_paises = {'La Liga':'SP1',
               'Bundesliga':'D1', 
               'Premier League':'E0',
              }
```

```python
# obtener un elemento del diccionario
dict_paises['La Liga']
```

```python
for liga in dict_paises:
    print(liga)
```

##### **Código**

```python
dict_historical_data = {} 

# looping through key elements
for liga in dict_paises:
    lista_df = [] # one dataframe per league
    for temporada in range(19, 24):
        df = pd.read_csv(base + str(temporada) + str(temporada+1) + "/" + dict_paises[liga]+ ".csv")
        df.insert(1, 'temporada', str(temporada)+"/"+str(temporada+1))
        lista_df.append(df)
    df_frames = pd.concat(lista_df) # almacenar todas las temporadas de una liga en un único dataframe
    dict_historical_data[liga] = df_frames  # asignar dataframe a un nuevo diccionario
```
